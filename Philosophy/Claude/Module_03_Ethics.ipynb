## What Is Ethics?

**Ethics** (from the Greek *ethos* = character, custom) is the branch of philosophy concerned with:

- **What is good?** — What things have value, and why?
- **What is right?** — Which actions are morally required, permitted, or forbidden?
- **How should we live?** — What kind of person ought I to be?
- **Why be moral at all?** — What is the foundation and authority of moral claims?

Ethics is not merely academic — it is inescapable. Every choice you make about how to treat others, every judgement you form about whether an action is fair or cruel, every political position you hold — all of these embed ethical commitments, whether examined or not. Philosophy's job is to examine them.

### The Three Levels of Ethics

| Level | Question | Example |
|-------|----------|---------|
| **Metaethics** | What *is* moral truth? Are moral facts real? | Is "killing is wrong" objectively true, or just an expression of feeling? |
| **Normative ethics** | What makes actions right or wrong? | Should I always tell the truth, or only when it produces good outcomes? |
| **Applied ethics** | How do we handle specific moral problems? | Is capital punishment justified? Is eating animals wrong? |

This module moves through all three levels, beginning with metaethics and working outward to the great normative theories before confronting some of the hardest applied problems.

# Chapter 1: Metaethics — Are Moral Facts Real?

---

## 🎯 The Core Question

When you say "slavery is wrong", are you:

- (a) Stating an **objective fact** about the world — something true independently of what anyone thinks or feels?
- (b) Expressing a **feeling or attitude** — something like "Boo, slavery!", which isn't true or false?
- (c) Issuing a **social norm** that happens to hold in your culture?
- (d) Making a claim that is **true relative to your framework** but not absolutely?

This is the central question of **metaethics**: not *what* is right or wrong, but *what kind of thing* moral claims even are.

---

## 📜 The Landscape: Moral Realism vs. Anti-Realism

### Moral Realism

**There are objective moral facts, independent of what any individual or culture believes.**

"Torturing children for fun is wrong" is as objectively true as "the Earth orbits the Sun" — not because we collectively agree to it, but because it is a genuine feature of moral reality.

**Varieties of moral realism:**

- **Naturalism:** Moral facts are natural facts, discoverable by science and reason. "Wrong" means something like "harmful to flourishing" or "disapproved of by ideally rational observers." (Mill, contemporary Cornell realists)
- **Non-naturalism:** Moral facts are real but not reducible to natural facts — they are a distinct category of reality, grasped by a special faculty of moral intuition. (G.E. Moore, W.D. Ross)
- **Moral rationalism:** Moral truths are knowable a priori by reason alone, just as mathematical truths are. (Kant)

### Moral Anti-Realism

**There are no objective moral facts — or if there are, we have no access to them.**

- **Expressivism / Emotivism (Ayer, Stevenson):** Moral statements don't express facts at all. "Murder is wrong" means something like "Murder — boo!" It is a non-cognitive expression of attitude, not a truth-apt claim.
- **Error Theory (Mackie):** Moral statements *do* try to state facts, but all such statements are false because objective moral values don't exist. Morality is a useful, perhaps necessary, fiction.
- **Relativism:** Moral truths are relative — either to individuals (subjectivism) or to cultures (cultural relativism). "X is wrong" means "X is disapproved of by me/my culture."
- **Quasi-realism (Blackburn):** We can be anti-realists about moral metaphysics while still using moral language in a fully engaged, realistic-sounding way. We *earn* the right to talk as if moral claims are true.

## 🧪 Thought Experiment 1: The Moral Progress Test

Most people believe that the abolition of slavery was **moral progress** — not just a *change* in moral views, but a genuine *improvement*.

Now consider:

- If moral relativism is true, then slave-owning societies weren't *wrong* — they were just operating by different cultural norms. Progress, then, is just *change*, not improvement. But this seems deeply unsatisfying.
- If expressivism is true, saying "slavery is wrong" is just venting a feeling. When abolitionists fought and died to end slavery, were they just acting out of especially intense feelings? Their sacrifice seems to demand something more.
- If moral realism is true, abolitionists *discovered* something true: that enslaved people have moral worth equal to their enslavers. This makes moral progress intelligible — like scientific progress.

**The moral progress argument for realism:** The fact that we can make sense of moral progress — that we can say, without irony, that we have learned that slavery is wrong — is powerful evidence that there are moral facts to discover.

**The anti-realist response:** We can explain moral progress without objective facts. Progress = increasing consistency, broadening the circle of empathy, reducing suffering. We don't need a Platonic realm of moral facts for this.

---

## 🔬 G.E. Moore's Open Question Argument

G.E. Moore (1903) posed a devastating challenge to **moral naturalism** — the attempt to define moral terms like "good" in terms of natural properties like "pleasure" or "survival advantage."

Suppose someone defines: **"Good" = "that which produces pleasure."**

Now ask: *"Is pleasure good?"*

If "good" just *means* "pleasure-producing," then this question becomes: *"Is pleasure pleasure?"* — which is trivially, boringly true. But the original question doesn't feel trivial. It feels like a real, open question.

This is Moore's **Open Question Argument**: for any natural property N you substitute for "good," the question "But is N actually *good*?" remains genuinely open. This shows that "good" cannot be identical to any natural property.

Moore called the error of defining moral terms in natural terms the **Naturalistic Fallacy**.

**Response:** Many philosophers think Moore's argument proves too much — or that it relies on a flawed theory of meaning. The debate continues.

---

## ⚖️ The Moral Relativism Debate

Cultural relativism is the most widely held view among non-philosophers — and one of the most thoroughly criticised by philosophers.

**The case for relativism:**
- Moral beliefs vary dramatically across cultures and history.
- Who are we to impose our values on others? Relativism seems tolerant and humble.
- There's no independent 'view from nowhere' from which to judge cultures.

**Devastating objections (James Rachels):**

1. **The tolerance argument backfires.** If morality is relative, then the claim "we ought to tolerate other cultures" is *itself* only relatively true. The relativist can't coherently ground tolerance.
2. **It rules out moral progress.** If a culture that endorsed slavery simply had *different* norms, abolitionists were not discovering anything — they were just being culturally deviant.
3. **Cultural variation doesn't prove relativism.** People disagree about empirical facts too — that doesn't make facts relative. Disagreement is evidence of *difficulty*, not of *no truth*.
4. **Some things seem wrong regardless of culture.** Torturing infants for entertainment, genocide, systemic rape — calling these "wrong only relative to our culture" is deeply implausible.

# Chapter 2: Consequentialism — The Ethics of Outcomes

---

## 🎯 The Core Claim

**The moral worth of an action is determined entirely by its consequences.**

What matters morally is not your intentions, your duties, or your character — it is the actual *outcomes* your action produces. An action is right if it produces the best available outcomes; wrong if it produces worse outcomes than an available alternative.

This family of views is called **consequentialism**. Its most famous member is **utilitarianism**.

---

## 🏛️ Classical Utilitarianism: Bentham and Mill

### Jeremy Bentham (1748–1832)

Bentham articulated the founding principle of utilitarianism:

> "Nature has placed mankind under the governance of two sovereign masters, **pain** and **pleasure**. It is for them alone to point out what we ought to do..."

His **Principle of Utility** (the Greatest Happiness Principle):

> **An action is right if and only if it produces the greatest happiness (pleasure minus pain) for the greatest number of people affected.**

Bentham proposed a **hedonic calculus** — a method for quantifying pleasures and pains according to their intensity, duration, certainty, propinquity (nearness), fecundity, purity, and extent (how many people are affected).

### John Stuart Mill (1806–1873)

Mill refined Bentham's view in several important ways:

1. **Quality of pleasures:** Not all pleasures are equal. *Higher* pleasures (intellectual, aesthetic) are intrinsically superior to *lower* pleasures (physical).

> "It is better to be Socrates dissatisfied than a fool satisfied. It is better to be a human being dissatisfied than a pig satisfied."

2. **Rule utilitarianism:** Rather than calculating consequences case by case, we should follow *rules* whose general acceptance produces the best outcomes. This explains why we should generally keep promises and tell the truth, even when breaking them might produce better results in a specific case.

3. **Liberalism:** Mill connected utilitarianism to his famous harm principle: the only justification for society's interference with an individual's liberty is to prevent harm to others.

## 🧪 Thought Experiment 2: The Trolley Problem

*Philippa Foot (1967), developed by Judith Jarvis Thomson*

**Case A — The Switch:**
A runaway trolley is heading toward five people tied to the tracks. They will certainly be killed. You are standing next to a lever. If you pull the lever, the trolley will be diverted to a side track, where only one person is tied. That person will be killed, but the five will be saved.

**Should you pull the lever?**

Most people say: **Yes.** Five lives saved, one lost — the numbers seem to justify it.

**The utilitarian calculation:** Pull the lever. Net lives saved = 4. It is the right action.

---

**Case B — The Footbridge:**
Same trolley, same five people. Now you are on a footbridge above the tracks. Standing next to you is a large stranger. If you push this person off the bridge, their body will stop the trolley, saving five lives. The stranger will die.

**Should you push the stranger?**

Most people say: **No** — this feels like murder, even if the arithmetic is identical (one death saves five).

**The utilitarian puzzle:** The numbers are the same. Why is one case permitted and the other not?

This divergence in intuitions is called the **trolley problem**. It reveals a deep tension between:
- Pure consequentialist reasoning (the numbers are all that matter), and
- A sense that there are constraints on *how* we may treat people — constraints the consequentialist struggles to explain.

The **Doctrine of Double Effect** (a deontological principle) attempts to explain the difference: in Case A, the one person's death is a *foreseen side effect*; in Case B, their death is the *intended means* to the good end. Intending to use someone as a means is morally different from foreseeing collateral harm.

## 🧪 Thought Experiment 3: The Experience Machine

*Robert Nozick, 1974*

Suppose scientists have developed an **Experience Machine**: a tank you can float in, connected to electrodes that stimulate your brain to give you any experiences you choose. You could experience writing a great novel, making close friends, living an exciting life. From the inside, it would be indistinguishable from reality.

**Would you plug in — permanently?**

Nozick argues: most people would say no — and this reveals that we don't *only* care about experiences (pleasures and pains). We also care about:
- **Actually doing things**, not just having the experience of doing them.
- **Being a certain kind of person** — brave, honest, loving — not just feeling like one.
- **Contact with a deeper reality** beyond the human-made simulation.

If the experience machine shows that pleasure isn't the *only* thing that matters, then **hedonistic utilitarianism is false**. There must be other intrinsic goods besides pleasure.

**Utilitarian responses:**
- **Preference utilitarianism (Peter Singer):** What matters is not pleasure per se but the satisfaction of *preferences*. The preference not to be deceived, the preference for genuine relationships — these are preferences the machine fails to satisfy.
- **Objective list theories:** Some things (knowledge, friendship, achievement) are good independently of whether they're desired or enjoyed.

---

## ⚔️ The Demanding Objection

Peter Singer's famous **drowning child argument** (1972):

> If you are walking past a shallow pond and see a child drowning, you should wade in and save them — even if it ruins your expensive new shoes. The cost to you (ruined shoes) is vastly outweighed by the benefit (a child's life).

Most people agree. But then Singer continues:

> Children are dying of preventable diseases in the developing world right now. You could save their lives by donating money. The moral situation is structurally identical to the drowning child. Therefore, you are morally required to donate until you reach the point of *marginal utility* — until giving more would harm you as much as it helps others.

**This is extraordinarily demanding.** It would mean giving away the vast majority of your income, foregoing luxuries, perhaps never buying new clothes while children are starving.

**Responses:**
- **Bite the bullet:** Singer accepts the demanding conclusion. Most of our ordinary behaviour is morally indefensible. We are implicated in preventable deaths by our inaction.
- **Agent-relative permissions:** We are morally permitted (not required) to give extra weight to our own projects, relationships, and commitments — morality can demand much, but not *everything*.
- **Rule utilitarianism:** A rule demanding total sacrifice is too psychologically demanding to be widely followed — a less demanding rule might produce better outcomes in practice.

````python
# ============================================================
# INTERACTIVE: Utilitarian Calculator
# Evaluate scenarios under different forms of utilitarianism
# ============================================================

def utilitarian_analysis(scenario, actions):
    """
    Evaluates a set of possible actions under:
    - Act utilitarianism (maximise net utility case by case)
    - Rule utilitarianism (follow the rule that, if generally adopted, maximises utility)
    - Preference utilitarianism (satisfy the most preferences, weighted by intensity)
    
    Each action dict has:
        name, act_utility, rule_utility, preference_satisfaction, notes
    """
    print(f"\n{'='*68}")
    print(f"SCENARIO: {scenario}")
    print(f"{'─'*68}")
    print(f"{'Action':<30} {'Act Util':>9} {'Rule Util':>10} {'Pref Sat':>9}")
    print(f"{'─'*68}")

    for a in actions:
        print(f"  {a['name']:<28} {a['act_utility']:>+9.1f} {a['rule_utility']:>+10.1f} {a['preference_satisfaction']:>+9.1f}")

    best_act  = max(actions, key=lambda x: x['act_utility'])
    best_rule = max(actions, key=lambda x: x['rule_utility'])
    best_pref = max(actions, key=lambda x: x['preference_satisfaction'])

    print(f"{'─'*68}")
    print(f"  Act utilitarianism recommends:        {best_act['name']}")
    print(f"  Rule utilitarianism recommends:       {best_rule['name']}")
    print(f"  Preference utilitarianism recommends: {best_pref['name']}")

    # Highlight conflicts
    names = {best_act['name'], best_rule['name'], best_pref['name']}
    if len(names) > 1:
        print(f"\n  ⚠️  THEORIES CONFLICT — different versions of utilitarianism disagree!")
    else:
        print(f"\n  ✅ All utilitarian theories agree.")

    print(f"\n  Notes:")
    for a in actions:
        print(f"    {a['name']}: {a['notes']}")


# --- Case 1: The Trolley Problem ---
utilitarian_analysis(
    scenario="The Trolley Problem: switch diverts trolley from 5 to 1.",
    actions=[
        {
            "name":                     "Pull the lever",
            "act_utility":              +4.0,   # 5 saved - 1 lost = net +4
            "rule_utility":             +3.5,   # Rule 'divert to save more' has good general effects
            "preference_satisfaction":  +4.0,   # More preferences satisfied
            "notes":                    "Standard utilitarian recommendation."
        },
        {
            "name":                     "Do nothing",
            "act_utility":              -5.0,   # 5 deaths
            "rule_utility":             -4.0,   # Rule 'never intervene' has poor general effects
            "preference_satisfaction":  -5.0,
            "notes":                    "Worst outcome on all utilitarian measures."
        },
    ]
)

# --- Case 2: Lying to a murderer ---
utilitarian_analysis(
    scenario="A murderer asks where your friend is hiding. Telling the truth leads to death.",
    actions=[
        {
            "name":                     "Lie to protect friend",
            "act_utility":              +8.0,   # Friend lives; small cost of deception
            "rule_utility":             +3.0,   # Rule 'lie to murderers' has okay effects overall
            "preference_satisfaction":  +8.0,   # Friend's preference to live is overriding
            "notes":                    "Act and preference utilitarianism strongly favour lying here."
        },
        {
            "name":                     "Tell the truth",
            "act_utility":              -9.0,   # Friend dies; murderer satisfied
            "rule_utility":             +5.0,   # Rule 'always tell truth' has broad social benefits
            "preference_satisfaction":  -9.0,
            "notes":                    "Rule utilitarianism may diverge — truth-telling rule has enormous social value."
        },
    ]
)

# --- Case 3: Singer's Giving Scenario ---
utilitarian_analysis(
    scenario="Singer: donate to effective charity vs. spend on luxury holiday.",
    actions=[
        {
            "name":                     "Donate to effective charity",
            "act_utility":              +90.0,  # Lives saved, suffering prevented
            "rule_utility":             +70.0,  # Rule 'donate significantly' has very good effects
            "preference_satisfaction":  +85.0,  # Many strong preferences satisfied
            "notes":                    "Singer: morally required, not merely supererogatory."
        },
        {
            "name":                     "Take the luxury holiday",
            "act_utility":              +5.0,   # Your pleasure; no one else benefits
            "rule_utility":             -10.0,  # Rule 'keep money for luxuries' has poor general effects
            "preference_satisfaction":  +5.0,   # Only your preferences count here
            "notes":                    "Much lower total utility — ordinary consumption hard to justify under utilitarianism."
        },
    ]
)

print("\n" + "="*68)
print("KEY INSIGHT: Act and Rule utilitarianism can conflict.")
print("Act: maximise this case. Rule: follow the best general policy.")
print("The lying-to-murderer case is Kant's most famous challenge to consequentialism.")
````

# Chapter 3: Deontology — The Ethics of Duty

---

## 🎯 The Core Claim

**Some actions are intrinsically right or wrong — independently of their consequences.**

Deontology (from the Greek *deon* = duty) holds that morality is fundamentally about **duties, rights, and rules** — constraints on how we may treat people that hold even when violating them would produce better outcomes.

The paradigm case: you must not torture an innocent person, even to save a thousand others. The wrongness of torture is not merely a matter of its bad consequences — it violates something about the moral status of the person being tortured.

---

## 🏛️ Kant's Moral Philosophy

Immanuel Kant (1724–1804) constructed the most systematic and influential deontological theory in the history of philosophy.

### The Good Will

Kant begins with a striking claim:

> "Nothing in the world — indeed nothing even beyond the world — can possibly be conceived which could be called good without qualification except a **good will**."

Intelligence, courage, wealth, even happiness — all of these can be used for evil. Only a good will — a will that acts from **duty** — is unconditionally good.

Crucially: what matters morally is not the *result* of your action but its *motive*. If you help a friend because you enjoy their company, that has moral worth — but it is the right action for the wrong reason. If you help a friend out of a sense of *duty* — because it is the right thing to do, even when you don't feel like it — *that* has full moral worth.

> Acting from duty, not merely in accordance with duty, is what gives an action moral worth.

### The Categorical Imperative

Kant argues that moral law must be **categorical** — unconditionally binding — not **hypothetical** ("do X *if* you want Y").

A hypothetical imperative: *"If you want to lose weight, exercise."* — only binding if you have that desire.

A categorical imperative: *"Do not lie."* — binding regardless of your desires or the consequences.

Kant formulates the **Categorical Imperative** in several ways, which he believes are equivalent:

#### Formula 1: The Universal Law Formula

> **"Act only according to that maxim whereby you can at the same time will that it should become a universal law."**

Before acting, identify the *maxim* (principle) of your action. Then ask: what if everyone acted on this maxim? Could you rationally will this to be universal?

**Example — lying to borrow money you can't repay:**
Maxim: "When I need money, I will make false promises to repay it."
Universal law: Everyone makes false promises when they need money.
**Result:** The very *institution* of promising collapses — no one would believe promises. The maxim *contradicts itself* when universalised. Therefore, lying is wrong.

#### Formula 2: The Humanity Formula

> **"Act so that you treat humanity, whether in your own person or in that of another, always as an end and never as a means only."**

Human beings have **dignity** — an absolute moral worth that makes them ends in themselves, not merely tools for others' purposes.

You may *use* people as means (e.g., a plumber fixes your pipes for pay), but you may not use them *merely* as means — treating their interests as irrelevant except insofar as they serve you.

This formula explains the deep wrongness of slavery, torture, manipulation, and using someone as a mere instrument: it violates their status as rational, self-governing beings.

## 🧪 Thought Experiment 4: Lying to the Murderer

*Kant's notorious case*

A murderer comes to your door and asks whether your friend — who they intend to kill — is hiding in your house. Your friend is, in fact, inside.

**Must you tell the truth?**

Kant's (in)famous answer: **Yes.** You must not lie, even to a murderer. Your duty to tell the truth is unconditional.

**Why?** Kant's argument:
1. Lying is wrong because the maxim of lying cannot be universalised without contradiction.
2. Moreover, if you lie and your friend sneaks out and is found and killed anyway — *you* bear some moral responsibility. If you told the truth, you are blameless for the consequences.
3. You are not responsible for what the murderer does with true information. You are responsible for your own actions.

**Responses:**
- Almost everyone finds this conclusion monstrous — a reductio ad absurdum of rigid rule-following.
- **W.D. Ross (1930):** We have *prima facie* duties (to tell truth, keep promises, prevent harm) that are generally binding but can be overridden by stronger duties in conflict cases. Here, the duty not to be complicit in murder overrides the duty to tell truth.
- **Kant's defenders:** The case reveals that what looks like a duty to lie is really a duty to refuse to play the murderer's game — perhaps by saying nothing, or "I won't tell you."

---

## 🏛️ W.D. Ross: Prima Facie Duties

Ross (1877–1971) offered a more flexible deontology. He argued that we have several **prima facie duties** — genuine moral obligations that are binding *unless overridden by a stronger duty in the circumstances*:

| Duty | Content |
|------|---------|
| **Fidelity** | Keep promises and commitments |
| **Reparation** | Make amends for past wrongs |
| **Gratitude** | Return benefits to benefactors |
| **Non-maleficence** | Do not harm others |
| **Beneficence** | Do good to others |
| **Self-improvement** | Improve your own virtue and knowledge |
| **Justice** | Distribute goods fairly |

When duties conflict (as they often do), we must exercise **moral judgement** to determine which is most pressing. No formula can replace this judgement — this is what separates mature ethical reasoning from mechanical rule-following.

---

## 🔬 Rights-Based Theories

A natural extension of deontology: if persons have dignity, they have **rights** — claims against others that constrain what may be done to them.

**Negative rights:** Rights not to be interfered with — to life, liberty, property. These correspond to duties of non-interference.

**Positive rights:** Rights to be provided with things — to education, healthcare, a fair trial. These correspond to duties to provide.

**Robert Nozick (1974) — Libertarian rights theory:**
People have absolute rights to self-ownership and to justly acquired property. Taxation for redistribution violates these rights — even if it produces better outcomes. Side constraints on action (never use people as mere tools) are not to be traded off against social benefits.

**John Rawls (1971) — Liberal rights theory:**
Justice requires principles that rational agents would choose from behind a **veil of ignorance** — not knowing their place in society, class, natural talents, or conception of the good. Rawls argues they would choose:
1. Equal basic liberties for all.
2. Social and economic inequalities only if they benefit the *least advantaged* members of society (the **Difference Principle**).

# Chapter 4: Virtue Ethics — The Ethics of Character

---

## 🎯 The Core Claim

**The fundamental question of ethics is not 'What should I do?' but 'What kind of person should I be?'**

Both consequentialism and deontology are **act-centred**: they ask what the right action is in a given situation. Virtue ethics is **agent-centred**: it focuses on the character of the moral agent — on the cultivation of stable dispositions (*virtues*) that reliably lead to good actions and constitute a good life.

Rather than asking "does this action maximise utility?" or "does it conform to duty?", virtue ethics asks: **"Is this what a person of good character would do?"**

---

## 🏛️ Aristotle's Virtue Ethics

Aristotle (384–322 BC) gave us the most fully developed virtue theory in his *Nicomachean Ethics* — still one of the most influential works in moral philosophy.

### Eudaimonia: The Good Life

Aristotle begins not with 'what is right?' but 'what is the good for humans?' He argues that every activity aims at some good. Medicine aims at health; architecture at beautiful buildings. What does human life as a whole aim at?

His answer: **eudaimonia** — often translated as "happiness" but better rendered as **flourishing** or *living well and doing well*. Eudaimonia is not a feeling (like pleasure) but an *activity* — a way of living and engaging with the world.

Crucially: eudaimonia is the *only* thing we seek for its own sake, as an ultimate end. Wealth, honour, pleasure — these are sought as *means* to flourishing, or as components of it. Eudaimonia is self-sufficient.

### The Function Argument

What is the *function* (*ergon*) of a human being? As a flute-player's function is to play the flute well, the human function must involve what is *distinctive* to humans.

What is distinctive to humans? **Rational activity** — living according to reason and excellence of character.

Therefore: **The good for humans = the active exercise of the soul in accordance with virtue.**

Eudaimonia is achieved by *doing things excellently* — not merely having good feelings or intentions.

### The Virtues

A **virtue** (*arete*) is an excellent trait of character — a stable disposition to feel and act in the right way, for the right reasons, at the right time.

Aristotle's key insight: **virtues are means between extremes (vices).**

| Deficiency (Vice) | Virtue (Mean) | Excess (Vice) |
|-------------------|--------------|---------------|
| Cowardice | **Courage** | Recklessness |
| Insensibility | **Temperance** | Self-indulgence |
| Miserliness | **Generosity** | Prodigality |
| Understatement | **Truthfulness** | Boastfulness |
| Surliness | **Friendliness** | Obsequiousness |
| Lack of spirit | **Righteous anger** | Irascibility |
| Humility (excessive) | **Magnanimity** | Vanity |

The mean is not a mathematical average — it is what is *appropriate* to the person, the situation, and the relationship. Finding the mean requires **practical wisdom** (*phronesis*).

### Practical Wisdom (Phronesis)

Practical wisdom is the master virtue — the ability to discern the right course of action in particular circumstances. It cannot be reduced to rules or algorithms. It requires:
- Experience in navigating human situations
- Sensitivity to the morally relevant features of a situation
- The capacity to deliberate well about what matters

This is why virtue ethics resists reduction to a simple formula: genuine moral expertise is more like the knowledge of a master craftsperson than the application of a rulebook.

### How Virtues Are Acquired

> "We become just by doing just acts, temperate by doing temperate acts, brave by doing brave acts."

Virtues are not innate — they are **habits** (*ethike* comes from *ethos* = habit). You become courageous by repeatedly acting courageously in situations of genuine risk. Character is formed by practice, by habituation, by the community and education that shapes you.

This is a profoundly practical vision of ethics: not a set of rules to memorise but a lifelong project of character formation.

## 🧪 Thought Experiment 5: The Mediocre Samaritan

Two people see a stranger fall in the street:

- **Person A** feels genuine compassion, wants to help, and immediately assists with warmth and care.
- **Person B** feels nothing — finds the situation annoying — but reasoning that it is their duty, helps anyway, stiffly and reluctantly.

**Who acted more morally?**

- **Kant says: Person B.** B acted from *duty*, not inclination. A's action has less moral worth because it was easy — motivated by pleasant feelings.
- **Aristotle says: Person A.** A's action flows from a cultivated virtue of compassion. B's reluctance reveals a *character flaw* — the virtuous person feels *appropriate emotions*, not just performs appropriate acts.
- **Utilitarians (mostly): Person A.** Both produce the same outcome, but A's emotional engagement makes them a *more reliable* helper in future cases.

This divergence is philosophically significant. Should morality *demand* that we overcome our inclinations? Or should it aim to *transform* our inclinations until virtue becomes second nature?

## ⚔️ Contemporary Virtue Ethics: MacIntyre and Narrative

**Alasdair MacIntyre** (*After Virtue*, 1981) argued that modern moral philosophy is in catastrophic disorder because it has abandoned the Aristotelian framework of human *telos* (purpose) that gives virtue its context.

Without a conception of *what humans are for*, moral discourse becomes fragmented — a series of incompatible claims with no shared rational basis. This explains why modern moral debates so often feel like irresolvable shouting matches.

MacIntyre's solution: we need to recover the concept of **practice** — socially complex activities with internal standards of excellence (chess, medicine, farming, science) — and of **narrative** — understanding the self not as an isolated atom but as a character in a story, embedded in traditions and communities.

Virtues are the qualities that sustain participation in practices and enable us to pursue the good of a whole human life. They are inseparable from community, tradition, and narrative identity.

---

## ⚔️ Care Ethics: Gilligan and Noddings

Carol Gilligan (1982) challenged the Kantian-Kohlberg tradition in moral psychology, arguing that it elevated an "ethic of justice" (abstract principles, impartiality, rights) over an **ethic of care** (relationships, responsiveness, particular attachments).

**Care ethics** holds that:
- Morality is rooted in **relationships and responsibilities** — not universal principles.
- **Particularity matters:** your special obligations to your child, your friend, your patient are morally primary, not merely instrumental to general duties.
- **Empathy and attentiveness** are core moral capacities — not biases to be overcome.

Care ethics is often grouped with virtue ethics because both centre character and emotional responsiveness rather than rule-following.

**Tension with impartiality:** The utilitarian calculates the welfare of strangers equally with loved ones. Care ethics resists this — our partial care for those close to us is not a moral failing but a moral *strength*. Whether these two views can be reconciled is an active debate.

## 🎯 The Core Question

How do we apply the frameworks of Chapters 2–4 to **real, contested moral problems**? And what happens when frameworks *disagree* — or when a single framework generates a repugnant conclusion?

Applied ethics does not mechanically deploy a single framework. Good applied ethical reasoning:
1. **Identifies the morally relevant features** of a situation (Who is affected? What rights are at stake? What consequences follow?).
2. **Considers multiple frameworks** — where they agree, we have strong grounds for a conclusion; where they disagree, we must weigh them.
3. **Takes moral intuitions seriously** — powerful intuitions that resist theoretical justification may indicate a flaw in the theory, not the intuition.
4. **Seeks reflective equilibrium** (Rawls) — moving back and forth between principles and intuitions, adjusting both until they cohere.

## 🧪 Thought Experiment 6: Thomson's Violinist

*Judith Jarvis Thomson, 1971 — on abortion rights*

You wake up connected to a famous unconscious violinist. A music society has kidnapped you because you are the only person with the right blood type to keep the violinist alive. If you disconnect yourself in the next nine months, the violinist will die. If you stay connected, he will recover.

**Are you morally required to stay connected?**

Most people say: **No.** You have a right to control what happens to your body. Even if the violinist has a right to life, that right doesn't include the right to use *your* body against your will.

Thomson's point: even *granting* that a foetus is a person with a right to life (a major concession), this does not automatically establish that abortion is wrong — because rights to life do not automatically override another person's right to bodily autonomy.

**Responses:**
- **Disanalogy arguments:** You didn't choose to be connected to the violinist. In most pregnancies (excepting rape), the woman's action contributed to the situation. Does responsibility for one's choices affect the extent of duty?
- **Special obligations:** Parents may have special obligations to their children that strangers don't have to violinists.
- **The minimally decent Samaritan:** Thomson distinguishes what we are *required* to do from what it would be *morally admirable* to do. Being a Good Samaritan is supererogatory — beyond duty.

## 🧪 Thought Experiment 7: The Repugnant Conclusion

*Derek Parfit, 1984 — Population Ethics*

Classical total utilitarianism says: we should maximise the total amount of wellbeing. Consider:

- **World A:** 10 billion people each living extremely well — flourishing, deeply happy lives.
- **World Z:** 100 trillion people each living lives barely worth living — just above zero on the scale of wellbeing.

If there are enough people in Z, their combined wellbeing exceeds that of A — making Z *morally better* than A according to total utilitarianism.

But this seems deeply **repugnant.** A world of vast numbers of barely-surviving people is surely *worse* than a world of flourishing billions.

Parfit called this the **Repugnant Conclusion** — and spent decades trying to find a theory of population ethics that avoids it. He died without finding a satisfactory solution. This remains one of the deepest unsolved problems in ethics.

**Attempted solutions:**
- **Average utilitarianism:** Maximise average wellbeing, not total. But this generates other repugnant conclusions (adding very happy people to a world can decrease the average).
- **Critical level utilitarianism:** Only count people whose wellbeing exceeds a critical threshold.
- **Person-affecting views:** An outcome is only better if it is better *for someone*. Non-existent people can't be harmed by not being created.

## ⚖️ Four Major Applied Debates

### 1. Animal Ethics

**The utilitarian case (Singer):** Pain is pain regardless of who suffers. If animals can suffer, then their suffering counts morally — equal to the suffering of humans of similar sentience. Factory farming causes immense suffering to billions of animals. This is one of the greatest moral catastrophes in human history, largely ignored.

**The Kantian response:** Kant held that we have no direct duties to animals (who lack rationality), only indirect duties (cruelty to animals brutalises us and leads to cruelty to humans). Most contemporary philosophers find this inadequate.

**Tom Regan's rights approach:** Animals that are 'subjects of a life' — with beliefs, desires, memory, emotions — have inherent value and rights. They cannot be treated as mere means to human ends.

---

### 2. Global Justice and Poverty

**Singer's argument revisited:** If we can prevent something bad without sacrificing anything of comparable moral importance, we should do it. Affluent people can prevent children dying of preventable diseases at little cost to themselves. Therefore, we are morally required to do so — on a massive scale.

**Thomas Pogge's structural argument:** Global poverty is not merely a failure of affluent people to give — it is actively caused by unjust global institutional structures (trade rules, debt, resource extraction) that wealthy nations design and maintain. This transforms the obligation from charity to *restitution* — a much stronger demand.

**Limits of obligation:** How demanding can morality be? What space does it leave for personal projects, relationships, and a distinctively human life?

---

### 3. Bioethics: Autonomy, Life, and Death

Key principles in bioethics (Beauchamp & Childress):
- **Autonomy:** Respect persons' right to make informed decisions about their own bodies.
- **Non-maleficence:** Do no harm.
- **Beneficence:** Act in the patient's best interest.
- **Justice:** Distribute healthcare resources fairly.

These principles often conflict:
- **Euthanasia:** Autonomy supports a patient's right to end their life; non-maleficence questions whether doctors should participate. What is the patient's 'best interest'?
- **Resource allocation:** A scarce organ should go to the patient who will benefit most (beneficence) or most needs it (justice) — but these may be different people.
- **Enhancement:** If a drug makes healthy people smarter, is that different from glasses correcting vision? What if only the wealthy can afford it?

### 4. AI and Emerging Technology Ethics

Rapidly developing technologies raise novel ethical questions:
- **Autonomous weapons:** If a drone decides to kill without human oversight, who is morally responsible?
- **AI consciousness:** If an AI system has morally relevant experiences, do we have obligations to it? How would we know?
- **Algorithmic bias:** If a hiring algorithm systematically disadvantages certain groups, who is responsible — the designer, the user, the algorithm?
- **Privacy and surveillance:** What rights to privacy do we retain? When is mass surveillance justified by security benefits?
- **Future generations:** Do we have obligations to people who will exist in the future? If so, climate change becomes a massive intergenerational injustice.

## 🔬 The Method of Reflective Equilibrium

John Rawls described the method of **reflective equilibrium** — the proper method of ethical reasoning:

1. Start with your **moral intuitions** — the considered judgements you feel most confident about (gratuitous cruelty is wrong; fairness matters).
2. Formulate **general principles** that systematise and explain these intuitions.
3. Test principles against *new* cases — where they generate counterintuitive verdicts, decide: does the principle need revising, or should the intuition be reconsidered?
4. Move back and forth, adjusting both, until they **cohere**.

This is not circular — it is the same method used in science (matching theory to data, adjusting both) and it is the only honest method for moral reasoning. Ethics does not proceed from a single self-evident axiom. It triangulates from many sources.

## 📌 Key Terms Glossary

| Term | Definition |
|------|------------|
| **Metaethics** | The philosophical study of the nature, status, and foundations of moral claims |
| **Moral realism** | The view that objective moral facts exist, independently of what anyone believes |
| **Expressivism** | Moral statements express attitudes, not facts; they are not truth-apt |
| **Error theory** | Moral statements try to state facts but are systematically false (Mackie) |
| **Naturalistic fallacy** | Moore: the mistake of defining 'good' in terms of natural properties |
| **Open question argument** | For any natural definition of 'good', 'But is N really good?' remains open |
| **Consequentialism** | The rightness of an action is determined entirely by its consequences |
| **Utilitarianism** | Maximise the greatest happiness (pleasure minus pain) for the greatest number |
| **Hedonic calculus** | Bentham's method for quantifying and comparing pleasures and pains |
| **Higher/lower pleasures** | Mill: intellectual pleasures are qualitatively superior to physical ones |
| **Experience Machine** | Nozick's challenge to hedonism: we care about more than subjective experience |
| **Deontology** | Ethics of duty: some actions are intrinsically right/wrong regardless of outcomes |
| **Categorical Imperative** | Kant's supreme moral principle; admits two main formulations |
| **Universal Law Formula** | Act only on maxims you could will to be universal laws |
| **Humanity Formula** | Always treat persons as ends in themselves, never merely as means |
| **Good will** | Kant: the only unconditionally good thing; a will that acts from duty |
| **Prima facie duty** | Ross: a genuine moral obligation that can be overridden by stronger duties |
| **Veil of ignorance** | Rawls: choose principles of justice without knowing your place in society |
| **Difference Principle** | Rawls: inequalities are just only if they benefit the least advantaged |
| **Virtue ethics** | Ethics centred on character — cultivating stable dispositions of excellence |
| **Eudaimonia** | Aristotle: human flourishing — living and acting well in accordance with virtue |
| **Doctrine of the Mean** | Virtue is the midpoint between two vicious extremes (deficiency and excess) |
| **Phronesis** | Practical wisdom: the master virtue — knowing how to act well in particular cases |
| **Supererogatory** | Going beyond what duty requires; morally praiseworthy but not obligatory |
| **Doctrine of Double Effect** | It can be permissible to cause harm as a foreseen side effect, but not as an intended means |
| **Reflective equilibrium** | Rawls: moral reasoning as coherence between intuitions and principles |
| **Repugnant Conclusion** | Parfit: total utilitarianism implies a vast population of barely-living people may be 'best' |
| **Care ethics** | Morality is rooted in relationships and responsiveness, not impartial principles |
