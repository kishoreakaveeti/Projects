## What Is Epistemology?

**Epistemology** (from the Greek *episteme* = knowledge, *logos* = study) is the branch of philosophy concerned with:

- **What is knowledge?** — What does it mean to truly *know* something?
- **What can we know?** — Are there limits to human knowledge?
- **How do we know?** — What are the valid *sources* of knowledge?
- **How certain can we be?** — What's the difference between *knowing* and *believing*?

It sounds abstract, but epistemology is astonishingly practical. Every time a scientist runs an experiment, every time a judge weighs evidence, every time you decide whether to trust a news article — you are implicitly doing epistemology.

# Chapter 1: The Problem of Knowledge & Justified Belief

---

## 🎯 The Core Question

What separates **knowing** something from merely **believing** it?

This seems trivial at first. But consider:

- A gambler *believes* their lucky number will win. And it does. Did they *know* it would win?
- A doctor *believes* a patient has pneumonia after checking symptoms. She is right. Did she *know* it?
- You *believe* the sun will rise tomorrow. It does. Did you *know* it would?

Intuitively, the gambler got *lucky*, the doctor made a *justified inference*, and you had a *well-grounded expectation*. But what exactly is the difference? That's what this chapter is about.

---

## 📜 The Classical Definition of Knowledge: JTB

The dominant answer for over 2,000 years — from Plato's *Theaetetus* onwards — is called the **Justified True Belief** (JTB) theory.

It states:

> **You know that P if and only if:**
> 1. **P is true** (the Truth condition)
> 2. **You believe that P** (the Belief condition)
> 3. **Your belief is justified** (the Justification condition)

Let's unpack each condition carefully.

---

### Condition 1: Truth

Knowledge requires that the thing you claim to know is actually *true*. You cannot know something false.

- You can *believe* that the Earth is flat. But you cannot *know* it, because it isn't flat.
- Medieval people believed the Sun orbited the Earth. They did not *know* this — they were wrong.

This seems obvious, but it has an important implication: **knowledge is always about reality**, not just about our mental states.

---

### Condition 2: Belief

You can only know something you actually believe. If someone asks you "What's the capital of Australia?" and you say "Sydney" but you genuinely believe the answer is "Canberra" — then you don't *know* it's Sydney, even if you said it correctly by accident.

Belief is a mental *commitment* to a proposition being true. It's different from:
- **Guessing** (no real commitment)
- **Suspecting** (partial commitment)
- **Hoping** (desire rather than commitment)

---

### Condition 3: Justification

This is the most philosophically interesting condition. **Justification** is what separates a *reasonable* belief from a *lucky guess*.

Consider two people who both believe it will rain tomorrow:
- **Person A** checked the weather app, looked at dark clouds, and noticed falling barometric pressure.
- **Person B** just *feels* it in their bones.

Both believe the same thing. It does rain. But we'd say Person A's belief was *justified* (grounded in evidence and good reasoning) while Person B's was not (it was a lucky guess or coincidence).

Justification means: **you have good reasons for believing what you believe**.

---

## 🧪 Thought Experiment 1: The Broken Clock

Imagine you have a clock on your wall. It stopped exactly 12 hours ago, so it now reads 3:00. You glance at it and think: "It's 3 o'clock."

And it IS 3 o'clock.

**Questions to consider:**
1. Is your belief TRUE? ✅ Yes — it really is 3 o'clock.
2. Do you BELIEVE it? ✅ Yes — you genuinely believe it's 3 o'clock.
3. Is your belief JUSTIFIED? 🤔 You looked at a clock — clocks are normally reliable sources of time...

**But do you KNOW it's 3 o'clock?**

Most people's intuition says: **No, not really.** You got lucky. Your evidence (the clock) was unreliable in a way you didn't know. This suggests that justification needs to be not just *present* but also *connected to the truth in the right way*.

This is the seed of what will become the **Gettier Problem** in Chapter 4.

````python
# ============================================================
# INTERACTIVE: The JTB Analyzer
# Test whether a belief counts as "knowledge" under JTB theory
# ============================================================

def analyze_jtb(scenario, is_true, is_believed, is_justified):
    """
    Analyzes whether a belief constitutes knowledge under the
    classical Justified True Belief (JTB) theory.
    """
    print(f"\n{'='*55}")
    print(f"SCENARIO: {scenario}")
    print(f"{'='*55}")
    print(f"  ✅ True?       {'YES' if is_true      else 'NO  ❌'}")
    print(f"  ✅ Believed?   {'YES' if is_believed  else 'NO  ❌'}")
    print(f"  ✅ Justified?  {'YES' if is_justified else 'NO  ❌'}")
    print(f"  {'─'*47}")
    
    is_knowledge = is_true and is_believed and is_justified
    if is_knowledge:
        print(f"  🎓 VERDICT: This IS knowledge (under JTB).")
    else:
        failed = []
        if not is_true:       failed.append("truth")
        if not is_believed:   failed.append("belief")
        if not is_justified:  failed.append("justification")
        print(f"  ❌ VERDICT: NOT knowledge. Missing: {', '.join(failed)}.")
    print()

# --- Try some scenarios ---

analyze_jtb(
    scenario   = "The gambler believes #7 will win. It does.",
    is_true      = True,
    is_believed  = True,
    is_justified = False   # No good reason to believe this
)

analyze_jtb(
    scenario   = "The doctor diagnoses pneumonia based on X-rays & symptoms. Correct.",
    is_true      = True,
    is_believed  = True,
    is_justified = True    # Medical evidence is good justification
)

analyze_jtb(
    scenario   = "You believe the Earth is flat (it isn't).",
    is_true      = False,
    is_believed  = True,
    is_justified = False
)

analyze_jtb(
    scenario   = "The broken clock shows 3:00. It happens to be 3:00.",
    is_true      = True,
    is_believed  = True,
    is_justified = True    # You had a reason (the clock) -- but was it GOOD reason?
)

print("NOTE: The broken clock case passes JTB but still feels wrong.")
print("This tension motivates the Gettier Problem in Chapter 4!")
````

## 🔗 Types of Justification

Philosophers disagree about what makes a belief *justified*. Here are the main positions:

### 1. Foundationalism
Some beliefs are **basic** — they don't need justification from other beliefs. They are self-evident, or directly grounded in experience. All other beliefs are justified by being built on top of these foundations.

- **Example basic beliefs:** "I am in pain right now." "2+2=4." "I seem to see a red object."
- **Analogy:** A building rests on foundations. The foundations don't need anything *below* them.
- **Famous foundationalist:** Descartes — who tried to find the one thing he *couldn't* doubt.

### 2. Coherentism
No belief is special. A belief is justified if it **coheres** (fits together consistently) with the rest of your belief system — like a web where every node supports every other.

- **Analogy:** A detective's case. No single piece of evidence is foundational. The whole picture hangs together.
- **Problem:** Two people can have internally coherent belief systems that contradict each other. Whose is "justified"?

### 3. Reliabilism
A belief is justified if it was produced by a **reliable process** — one that tends to produce true beliefs.

- **Example:** Perception, memory, mathematical reasoning are generally reliable. Horoscopes and wishful thinking are not.
- **Advantage:** This explains why the broken clock fails — even though you followed normal procedure (looking at a clock), the *process* was unreliable in this instance.

# Chapter 2: Rationalism vs. Empiricism

---

## 🎯 The Core Question

Where does knowledge ultimately **come from**?

Is knowledge derived from:
- **Reason alone** — through pure thought, logic, and reflection (Rationalism), or
- **Experience** — through the senses, observation, and experiment (Empiricism)?

This is one of the oldest and most productive debates in all of philosophy. It set the stage for the Scientific Revolution, Kant's critical philosophy, and modern cognitive science.

---

## 🧠 Rationalism: The Power of Pure Reason

**Core claim:** Some knowledge is **a priori** — knowable *prior to* and *independent of* sensory experience, through reason alone.

### Key Rationalists

| Philosopher | Dates | Key Idea |
|-------------|-------|----------|
| **Plato** | 428–348 BC | Knowledge is recollection of eternal Forms |
| **René Descartes** | 1596–1650 | Clear and distinct ideas grasped by reason are certain |
| **Baruch Spinoza** | 1632–1677 | Geometric method: deduce all from first principles |
| **Gottfried Leibniz** | 1646–1716 | Innate ideas; truths of reason vs. truths of fact |

### The Rationalist Case

Consider these statements:

1. **"2 + 2 = 4"** — Is this true because you counted 2 apples and 2 apples a thousand times? Or is it *necessarily* true, true in every possible world, grasped by reason alone?

2. **"All bachelors are unmarried"** — You don't need to survey all bachelors. The truth follows from the *meanings of the words*. Reason alone delivers this.

3. **"If A > B and B > C, then A > C"** — Pure logic. No observation needed.

Rationalists argue that the **most reliable and universal** knowledge — mathematics, logic, geometry, and certain metaphysical truths — comes from reason, not the senses.

### Descartes' Method

Descartes wanted to rebuild knowledge on an unshakeable foundation. His method was **radical doubt**: systematically reject everything that could possibly be doubted. What survives?

He doubted his senses (they sometimes deceive). He doubted the external world (he might be dreaming). He doubted mathematics (an evil demon might be making him wrong).

What he couldn't doubt: **the act of doubting itself**. If he's doubting, he must be *thinking*. If he's thinking, he must *exist*.

> ***"Cogito, ergo sum"*** — I think, therefore I am.

This is knowledge gained by **reason alone** — not by any observation.

---

## 🧠 Empiricism: The Primacy of Experience

**Core claim:** All knowledge ultimately derives from **sensory experience**. The mind begins as a blank slate (*tabula rasa*), and experience writes on it.

### Key Empiricists

| Philosopher | Dates | Key Idea |
|-------------|-------|----------|
| **John Locke** | 1632–1704 | Mind as tabula rasa; all ideas from experience |
| **George Berkeley** | 1685–1753 | To be is to be perceived (*esse est percipi*) |
| **David Hume** | 1711–1776 | Radical: even causation is just habitual expectation |

### The Empiricist Case

**Locke's Argument:** What does a person born blind know about colour? Nothing. Because all knowledge of colour comes from *seeing*. If we had no senses at all, our minds would be completely empty. Therefore, all content of the mind originates in experience.

**Hume's Radical Empiricism:** Hume pushed this further. He argued that even our idea of **causation** — the backbone of science — is not something we *observe*. When a billiard ball strikes another, we see one ball move, a collision, then the other ball move. We never *see* causation itself. We just expect it from habit. This is a shocking conclusion: causation might be a psychological habit, not a feature of reality.

Hume's challenge: *"If you cannot trace an idea back to a sensory impression, the idea is meaningless."* This became the foundation of modern scientific skepticism.

---

## ⚔️ A Priori vs. A Posteriori Knowledge

This is the key battleground:

| Term | Meaning | Examples |
|------|---------|----------|
| **A priori** | Knowable *before* / *independent of* experience | Math, logic, definitions |
| **A posteriori** | Knowable only *through* experience | "Water boils at 100°C", "Sydney is in Australia" |

A second distinction (from Kant):

| Term | Meaning | Examples |
|------|---------|----------|
| **Analytic** | True by *definition* — the predicate is contained in the subject | "Bachelors are unmarried", "Triangles have 3 sides" |
| **Synthetic** | Adds *new information* not contained in the definition | "The cat is on the mat", "Water boils at 100°C" |

### Kant's Bombshell: Synthetic A Priori Knowledge

Kant asked: Is *all* a priori knowledge just analytic (true by definition)? Or can there be **synthetic a priori** knowledge — knowledge that is both:
- Independent of experience (a priori), AND
- Genuinely informative / not just definitional (synthetic)?

His answer: **YES.** Examples:
- "7 + 5 = 12" — This is a priori (we don't need to count physical objects), but it's synthetic ("12" is not contained in the *definition* of "7 + 5"; it requires a mental step).
- "Every event has a cause" — This structures our entire experience, but it's not something we read off from experience.

Kant argued that the **structure of human cognition** (space, time, causation) is built in — it's what makes experience possible in the first place. This was a synthesis of rationalism and empiricism: experience *triggers* knowledge, but reason *structures* it.

> Kant said Hume "woke him from his dogmatic slumber" — and he spent the rest of his career answering Hume's challenge.

# Chapter 3: Skepticism — The Evil Demon & Brain-in-a-Vat

---

## 🎯 The Core Question

Can we be **certain** of *anything*? Is there any knowledge that is absolutely secure against all possible doubt?

Skepticism is the philosophical position that **knowledge is impossible** — or at least severely limited. Skeptical arguments have a remarkable structure: they don't claim the world doesn't exist; they just show that *our evidence* for the world is less solid than we assume.

---

## 📜 The History of Skepticism

Skepticism has ancient roots:

- **Pyrrho of Elis (c. 360–270 BC):** The founder of ancient skepticism. Argued that for every argument, an equally strong counter-argument can be made. We should therefore *suspend judgment* (*epoché*) on all things.
- **The Academics (Plato's school):** Argued that we can have *probable* beliefs but never *certain* knowledge.
- **Descartes (1641):** Used skepticism as a *method* — not as a final position, but as a tool to find what's truly certain.

---

## 🧪 Thought Experiment 3: The Dreaming Argument

Right now, as you read this, how do you know you are not **dreaming**?

Think carefully. When you dream, you are often convinced the dream is real. You interact with objects, have conversations, feel emotions. The dream seems utterly convincing *from the inside*. Only when you wake up do you know it was a dream.

But if dreams are convincing *while you're in them*, what test can you apply *right now* to verify that *this* experience isn't a dream?

- "I'll pinch myself!" — But you can pinch yourself in dreams.
- "Dreams are blurry and incoherent!" — Some dreams are very vivid and logical.
- "I remember waking up this morning!" — But memories can be fabricated within a dream.

Descartes concluded: **the dreaming argument casts doubt on all sensory knowledge** — we cannot be certain that our current experience represents reality.

---

## 🧪 Thought Experiment 4: The Evil Demon

*From Descartes' Meditations on First Philosophy (1641)*

Descartes went further than the dreaming argument. He imagined:

> "Suppose there exists an infinitely powerful, infinitely clever **Evil Demon** whose sole purpose is to deceive me. This demon makes me see a world, feel sensations, do mathematics — all of it perfectly crafted illusions. The sky, the ground, my own body — all fake. Even when I think 2+2=4, the demon makes me believe this false answer."

If such a demon *could* exist, then **nothing we believe** is safe. Not our memories, not our reasoning, not our perception, not even our mathematics.

**What survives the Evil Demon?**

Only the *Cogito* — the act of thinking/doubting itself. Even if the demon deceives me about everything, *there must be a me to deceive*. The Evil Demon cannot make it false that I am having experiences, because the very act of deception requires a subject to be deceived.

This is why Descartes calls the *Cogito* the **Archimedean point** — the one fixed point from which he can rebuild all knowledge.

---

## 🧪 Thought Experiment 5: The Brain in a Vat

*A modern version of Descartes' Evil Demon, by Hilary Putnam (1981)*

Imagine that your brain has been removed from your body and placed in a vat of nutrients. Mad scientists have connected your brain to a supercomputer that sends it exactly the same electrical signals that a normal brain would receive from a body living a normal life.

You would *believe* you are reading this notebook. You would *believe* you are in a room. You would have all the same experiences as you do now.

**Question:** Are you a brain in a vat?

**You cannot rule it out.** Everything you experience would be identical either way.

This scenario is familiar from popular culture:
- **The Matrix (1999):** Humans live in a simulation; their brains receive false inputs.
- **Inception:** Nested dream worlds; reality is indistinguishable from illusion.

---

## ⚔️ Responses to Skepticism

Philosophers have not just accepted skepticism — they've fought back. Here are the main responses:

### 1. Descartes' Own Response: God as Guarantor
Descartes tried to prove the existence of a non-deceptive God, who guarantees that our clear and distinct rational perceptions are true. This is widely considered unsuccessful — it involves circular reasoning (using reason to prove the reliability of reason).

### 2. Moore's Common Sense Reply (G.E. Moore, 1939)
Moore held up his hand and said: "Here is one hand." His point: we are *more* certain that we have hands than we are that any philosophical skeptical argument is valid. If a philosophical argument concludes "you don't know you have a hand," the right response is to reject the argument, not to reject the hand.

> "The certainty of common sense facts outweighs the certainty of philosophical premises."

### 3. Wittgenstein: On Certainty
Wittgenstein argued that doubt only makes sense *within* a framework of certainties. To doubt whether you have a hand, you already need to take a language, a world, an inquiry to be *real*. Skepticism destroys the very conditions for meaning.

> "Doubt comes after belief." Radical doubt is not a coherent starting point.

### 4. Putnam's Semantic Response to Brain-in-a-Vat
Putnam made a clever argument: if you *were* a brain in a vat, the word "brain" in your language would refer to simulated brains, and "vat" would refer to simulated vats. So when you say "I might be a brain in a vat," you'd be saying "I might be a simulated-brain in a simulated-vat" — which might be true! But it doesn't challenge external reality in the way you thought.

This is a subtle and controversial argument — but it shows how language and reference can bear on skeptical scenarios.

### 5. Pragmatism (William James, Charles Peirce)
Even if skepticism can't be refuted, **it doesn't matter**. What matters is whether our beliefs *work* — whether they help us navigate the world successfully. The Evil Demon hypothesis makes no practical difference to how we should live. So we should act as if the world is real.

---

## 🔍 Global vs. Local Skepticism

It's important to distinguish:

| Type | What is doubted | Example |
|------|----------------|---------|
| **Global skepticism** | *Everything* — the whole external world | Evil Demon, Brain-in-a-Vat |
| **Local skepticism** | *Specific domains* | Skepticism about memory, about other minds, about moral facts |

Local skepticism is much more common and defensible. You might be skeptical about astrology (local) without doubting that the physical world exists (global).

# Chapter 4: The Gettier Problem — What *Really* Is Knowledge?

---

## 🎯 The Core Question

We established that knowledge = Justified True Belief. But in 1963, Edmund Gettier published a 3-page paper that shook epistemology to its core. He showed that **JTB is not sufficient for knowledge** — you can have all three conditions and still not have knowledge.

> Edmund Gettier (1927–2021) spent his career resisting publication. His three-page 1963 paper "Is Justified True Belief Knowledge?" is one of the most influential papers in 20th century philosophy.

---

## 🧪 Gettier Case 1: The Job Applicant

*Original from Gettier (1963)*

Smith and Jones are applying for the same job. Smith has strong evidence for two things:
1. Jones will get the job.
2. Jones has 10 coins in his pocket (he just counted them).

From these, Smith infers:

> **"The man who will get the job has 10 coins in his pocket."**

This belief is *justified* (he has good evidence for both premises) and he is *committed* to it.

Now, as it turns out:
- Smith himself gets the job (not Jones).
- Smith also has 10 coins in his own pocket (he didn't know this).

So Smith's belief — **"the man who will get the job has 10 coins in his pocket"** — is **TRUE**.

Let's check JTB:
- ✅ **True?** Yes — Smith has the job and has 10 coins.
- ✅ **Believed?** Yes — Smith genuinely believed this.
- ✅ **Justified?** Yes — he had good evidence (Jones' coins, Jones getting the job).

JTB says this is **knowledge**. But intuitively — **it isn't!** Smith believed it for the wrong reasons. He got lucky. He thought it was about *Jones*, but it turned out to be true because of *himself*.

---

## 🧪 Gettier Case 2: The Sheep in the Field

*A famous variant by Roderick Chisholm*

You're driving through the countryside. You look at a field and see what looks exactly like a sheep. You form the belief:

> **"There is a sheep in that field."**

Your justification: you can clearly see a sheep-shaped thing in the field.

What you don't know: what you're actually looking at is a large rock that is exactly the shape and colour of a sheep. However, hidden behind a hill in the same field, *there is an actual sheep*.

Your belief is:
- ✅ **True** — there is indeed a sheep in the field (just not the one you can see).
- ✅ **Believed** — you believe it.
- ✅ **Justified** — your visual evidence is normally reliable.

But again — **you don't know it**. You got lucky. Your justification was for a false belief (that there's a sheep where you're looking), and the true belief was accidentally made true by something you were completely unaware of.

## 🧪 Gettier Case 3: The Clock Revisited

Recall from Chapter 1: the stopped clock that happens to show the right time. This is a Gettier case.

You look at the clock (which stopped 12 hours ago). It reads 3:00. It happens to be exactly 3:00.

- ✅ True: It is 3:00.
- ✅ Believed: You believe it's 3:00.
- ✅ Justified: You used a normally reliable source (a clock).

But you don't **know** it's 3:00. The truth was accidental relative to your justification.

---

## ⚔️ The Structure of a Gettier Case

Every Gettier case has the same structure:

1. You have **justified** belief in proposition **Q** (Jones gets the job, you see a sheep, the clock shows 3:00).
2. From Q, you infer (or directly believe) **P** (the person with 10 coins gets the job, there's a sheep in the field, it's 3:00).
3. **Q is false** (Jones doesn't get the job; that's not a sheep; the clock is broken).
4. But **P is *coincidentally* true** for a completely different reason.

The result: Justified True Belief *without* knowledge.

---

## 🔧 Responses: Fixing the JTB Definition

Philosophers scrambled to patch the definition. Here are the main attempts:

### 1. The No-False-Lemmas Condition
Add a fourth condition: **your justification must not essentially rely on any false beliefs.**

- Smith's error: his justification relied on the false belief that *Jones* would get the job.
- In the sheep case: relied on the false belief that *this object* is a sheep.

**Problem:** Some Gettier cases are constructed where no false lemma is used — making this patch incomplete.

### 2. The Causal Theory (Alvin Goldman, 1967)
Add: **your belief must be caused by the fact that makes it true**.

- In the sheep case: your belief is caused by a *rock* (false cause), not by the *actual sheep* (the fact that makes it true).
- If you had seen the actual sheep, that would have *caused* your belief appropriately.

**Problem:** How do we handle mathematical or historical knowledge? There's no obvious causal chain between, say, Caesar's death and my belief about it.

### 3. Reliabilism (Goldman, 1979)
A belief is knowledge if it is produced by a **reliable cognitive process** — one that produces true beliefs in normal conditions.

- A stopped clock is not a reliable process for telling time.
- Normal vision is reliable; hallucinations are not.

**Problem:** What counts as "reliable"? And reliable *in what circumstances*? (A stopped clock is reliable twice a day...)

### 4. Sensitivity / Safety Conditions (Nozick, Sosa)
**Sensitivity:** If P were false, you wouldn't believe it. (Counterfactual condition)

- If the clock weren't showing 3:00, would you still believe it's 3:00? Yes — because you'd look at the same broken clock showing 3:00. So you fail sensitivity.

**Safety:** Your belief couldn't *easily* have been wrong — in nearby possible worlds where conditions are similar, you'd still be right.

### 5. Contextualism (David Lewis, Keith DeRose)
"Knowledge" is **context-sensitive**. What counts as knowledge in everyday conversation is different from what counts in a philosophy seminar.

- In normal life: "I know where my keys are" = I have a reliable belief about their location.
- In a philosophy class discussing skepticism: the standards are much higher.

The **standards** for knowledge shift depending on the stakes and conversational context. This explains why the skeptic's scenarios feel compelling in philosophy class but irrelevant in daily life.

---

## 🌀 The Big Picture: Knowledge After Gettier

Despite 60 years of effort, **no universally accepted definition of knowledge** has emerged since Gettier. This has led some philosophers to wonder:

- Maybe knowledge is a **natural kind** — like "water" or "gold" — whose nature is discovered empirically rather than defined analytically.
- Maybe we should abandon the project of *defining* knowledge and instead study it **experimentally** (experimental philosophy / X-Phi) — asking ordinary people about cases and mapping our actual concept of knowledge.
- Maybe knowledge is **primitive and indefinable** — a basic concept that cannot be reduced to other concepts.

The Gettier problem is not just a puzzle — it reveals something deep: **ordinary concepts can resist precise philosophical analysis**.

````python
# ============================================================
# INTERACTIVE: Gettier Case Analyzer
# Test cases against different theories of knowledge
# ============================================================

def analyze_gettier_case(case_name, is_true, is_believed, is_justified,
                          no_false_lemma, causally_connected, reliable_process, 
                          sensitivity, description):
    """
    Analyzes a case against multiple theories of knowledge.
    """
    print(f"\n{'='*60}")
    print(f"🔍 CASE: {case_name}")
    print(f"{'─'*60}")
    print(f"   {description}")
    print(f"{'─'*60}")
    
    theories = {
        "Classical JTB":         is_true and is_believed and is_justified,
        "JTB + No False Lemma":  is_true and is_believed and is_justified and no_false_lemma,
        "Causal Theory":         is_true and is_believed and is_justified and causally_connected,
        "Reliabilism":           is_true and is_believed and reliable_process,
        "Sensitivity Account":   is_true and is_believed and is_justified and sensitivity,
    }
    
    print("   VERDICT BY THEORY:")
    for theory, counts_as_knowledge in theories.items():
        icon = "✅ KNOWLEDGE" if counts_as_knowledge else "❌ NOT knowledge"
        print(f"     {theory:<30} → {icon}")


analyze_gettier_case(
    case_name          = "The Stopped Clock (3:00pm)",
    description        = "You check a broken clock that reads 3:00. It happens to be exactly 3:00.",
    is_true            = True,
    is_believed        = True,
    is_justified       = True,   # Clocks are normally reliable sources
    no_false_lemma     = False,  # You rely on false premise: 'this clock is working'
    causally_connected = False,  # The fact (3:00pm) didn't cause your belief; the broken clock did
    reliable_process   = False,  # A broken clock is not a reliable time-telling process
    sensitivity        = False   # If it weren't 3:00, you'd still believe it (clock is stuck at 3:00)
)

analyze_gettier_case(
    case_name          = "The Sheep in the Field",
    description        = "You see a sheep-shaped rock. A real sheep is hidden behind a hill.",
    is_true            = True,
    is_believed        = True,
    is_justified       = True,   # Your eyes normally work fine
    no_false_lemma     = False,  # Relies on: 'that shape IS a sheep' (false)
    causally_connected = False,  # The actual sheep didn't cause your belief; the rock did
    reliable_process   = False,  # This particular process (seeing the rock) was unreliable
    sensitivity        = False   # If there were no sheep, you'd still believe it (you see the rock)
)

analyze_gettier_case(
    case_name          = "Normal Perception: You see a cat",
    description        = "You see a cat sitting on a mat. There really is a cat on the mat.",
    is_true            = True,
    is_believed        = True,
    is_justified       = True,
    no_false_lemma     = True,   # No false intermediary
    causally_connected = True,   # The cat's presence causes your visual experience
    reliable_process   = True,   # Normal vision is reliable
    sensitivity        = True    # If there were no cat, you wouldn't see one (no cat = no belief)
)

print("\n" + "="*60)
print("KEY INSIGHT: Gettier cases are ones where JTB is satisfied but")
print("all post-Gettier theories correctly deny knowledge — because the")
print("connection between justification and truth is accidental / lucky.")
````

## 📌 Key Terms Glossary

| Term | Definition |
|------|------------|
| **Epistemology** | The philosophical study of knowledge, belief, and justification |
| **Justified True Belief (JTB)** | The classical definition: knowledge = true + believed + justified |
| **A priori** | Knowledge independent of experience (math, logic) |
| **A posteriori** | Knowledge derived from experience (empirical facts) |
| **Analytic** | True by definition ("bachelors are unmarried") |
| **Synthetic** | True by adding new information ("water boils at 100°C") |
| **Rationalism** | The view that reason is the primary source of knowledge |
| **Empiricism** | The view that experience is the primary source of knowledge |
| **Skepticism** | The view that knowledge is impossible or severely limited |
| **Cogito** | Descartes' "I think, therefore I am" — the bedrock of certainty |
| **Gettier problem** | Counterexamples showing JTB is not sufficient for knowledge |
| **Reliabilism** | Knowledge = belief produced by a reliable cognitive process |
| **Contextualism** | The standards for knowledge vary with conversational context |
| **Tabula rasa** | Locke's "blank slate" — the mind before experience writes on it |
| **Synthetic a priori** | Kant: knowledge that is both independent of experience AND informative |
