# 15. Object-oriented programming in Python

```python
class StudySession:
    def __init__(self, topic, hours):
        self.topic = topic
        self.hours = hours

    def summary(self):
        return f"Studied {self.topic} for {self.hours} hours."

session = StudySession("linear algebra", 2.5)
print(session.topic)
print(session.summary())
```

```python
class ResearchSession(StudySession):
    def __init__(self, topic, hours, paper_name):
        super().__init__(topic, hours)
        self.paper_name = paper_name

    def summary(self):
        return f"Studied {self.topic} for {self.hours} hours while reading '{self.paper_name}'."

rs = ResearchSession("interpretability", 3.0, "Geometry of Truth")
print(rs.summary())
```

# 20. Automatic differentiation with PyTorch

Suppose
\[
f(x) = x^2 + 3x + 1
\]
Then
\[
\frac{df}{dx} = 2x + 3.
\]

```python
x = torch.tensor(2.0, requires_grad=True)

f = x**2 + 3*x + 1
f.backward()

print("x =", x.item())
print("f(x) =", f.item())
print("df/dx =", x.grad.item())
```

At \(x=2\), the derivative should be
\[
2(2) + 3 = 7.
\]

This is the engine behind backpropagation.

# 21. Writing cleaner Python

```python
def compute_mean(values):
    """Return the arithmetic mean of a non-empty list of numbers."""
    if not values:
        raise ValueError("values must not be empty")
    return sum(values) / len(values)

print(compute_mean([10, 20, 30]))
```

Common style conventions:
- `snake_case` for variables and functions
- `PascalCase` for classes
- `UPPER_CASE` for constants
