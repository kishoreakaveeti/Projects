# 1. The Core Idea: Information as Reduction in Uncertainty

The central intuitive idea of information theory is:

> **Information is what reduces uncertainty.**

Suppose you are waiting for the result of a coin toss.

Before you see the result, you are uncertain.

After someone tells you "Heads", your uncertainty is reduced.

That reduction in uncertainty is information.

Now imagine two kinds of messages:

- "The sun will rise tomorrow."
- "A fair coin landed heads."

The first message usually gives very little information because it was already very predictable.

The second gives more information because there was genuine uncertainty.

So information should depend on **how surprising** an event is.

This immediately suggests a relationship:

- common / predictable events $\rightarrow$ low information
- rare / surprising events $\rightarrow$ high information

This idea will lead us to the formula for **self-information**.

# 2. Self-Information

Let an event $x$ happen with probability $P(x)$.

The **self-information** (also called **surprisal**) of event $x$ is defined as

$$
I(x) = - \log_2 P(x)
$$

This quantity answers the question:

> "How much information do I gain when I learn that event $x$ occurred?"

---

## Why this formula makes sense

We want a function of probability that behaves in a reasonable way.

### Requirement 1: Rare events should carry more information

If an event is very unlikely, learning that it occurred should be surprising, so the information should be larger.

If $P(x)$ decreases, then $-\log P(x)$ increases.

Good.

### Requirement 2: Independent pieces of information should add

Suppose two independent events happen:

- event $x$ with probability $P(x)$
- event $y$ with probability $P(y)$

Then

$$
P(x, y) = P(x)P(y)
$$

We would like the information in learning both to be the sum of their separate informations:

$$
I(x, y) = I(x) + I(y)
$$

Now if

$$
I(x) = -\log P(x),
$$

then

$$
I(x, y) = -\log(P(x)P(y)) = -\log P(x) - \log P(y)
$$

which gives exactly the additive property we want.

That is one of the deepest reasons the logarithm appears in information theory.

---

## Some examples

### Example 1: Fair coin toss

If $P(x) = 0.5$, then

$$
I(x) = -\log_2(0.5) = 1
$$

So a fair coin outcome contains **1 bit** of information.

### Example 2: Event with probability $1/4$

$$
I(x) = -\log_2\left(\frac{1}{4}\right) = 2
$$

That event carries **2 bits** of information.

### Example 3: Certain event

If $P(x)=1$, then

$$
I(x) = -\log_2(1)=0
$$

A certain event carries **zero information** because it tells you nothing new.

---

## Important intuition

Self-information is not the same as "importance" in the everyday sense.

A surprising typo in a text message may have high self-information mathematically, even if it is not important in life.

In information theory, "information" has a very precise mathematical meaning tied to **uncertainty and probability**, not human meaning or value.

```
import numpy as np

probs = [1.0, 0.5, 0.25, 0.125, 0.01]

print("Self-information examples:")
for p in probs:
    info_bits = -np.log2(p)
    print(f"P(x) = {p:>5}  ->  I(x) = {-np.log2(p):.4f} bits")
```

# 3. Units: Bits, Nats, and Logarithm Bases

The base of the logarithm determines the unit:

- base 2 $\rightarrow$ **bits**
- base $e$ $\rightarrow$ **nats**
- base 10 $\rightarrow$ **hartleys** (rarely used in ML)

If we use base 2:

$$
I(x) = -\log_2 P(x)
$$

If we use natural log:

$$
I(x) = -\ln P(x)
$$

Both are valid. They differ only by a constant scaling factor.

Using the identity

$$
\log_2 x = \frac{\ln x}{\ln 2},
$$

we see that bits and nats carry the same conceptual content; only the unit changes.

In machine learning, many derivations use natural logarithms because calculus is cleaner. So loss functions may effectively be measured in **nats** rather than bits.

That is completely fine.

# 4. From Single Event Information to Average Information: Entropy

Self-information tells us how much information is carried by **one specific event**.

But often we want to know:

> On average, how much information does a random variable produce?

That average quantity is **entropy**.

For a discrete random variable $X$ with possible values $x$ and probabilities $P(x)$, the entropy is

$$
H(X) = - \sum_x P(x)\log_2 P(x)
$$

This is just the expected value of self-information:

$$
H(X) = \mathbb{E}[I(X)] = \mathbb{E}[-\log_2 P(X)]
$$

---

## Interpretation of entropy

Entropy measures the average uncertainty of a random variable.

- If the outcome is almost certain, entropy is low.
- If many outcomes are possible and equally likely, entropy is high.

---

## Examples

### Example 1: Certain outcome

Suppose $X$ always equals one value with probability 1.

Then

$$
H(X) = -1 \cdot \log_2(1)=0
$$

No uncertainty, no entropy.

### Example 2: Fair coin

For $P(H)=0.5$, $P(T)=0.5$:

$$
H(X) = -0.5\log_2(0.5)-0.5\log_2(0.5)=1
$$

So a fair coin has entropy **1 bit**.

### Example 3: Biased coin

Let $P(H)=0.9$, $P(T)=0.1$.

Then

$$
H(X) = -0.9\log_2(0.9) - 0.1\log_2(0.1)
$$

This is less than 1 bit, because the outcome is more predictable than a fair coin.

---

## Very important distinction

Entropy is not "disorder" in the vague everyday sense here.

In information theory, entropy means:

> **Average uncertainty**  
or  
> **Average information content**

That is the meaning to hold onto.

```
import numpy as np

def entropy(p):
    p = np.array(p, dtype=float)
    p = p[p > 0]  # ignore zero-probability terms to avoid log(0)
    return -np.sum(p * np.log2(p))

examples = {
    "Certain event [1.0]": [1.0],
    "Fair coin [0.5, 0.5]": [0.5, 0.5],
    "Biased coin [0.9, 0.1]": [0.9, 0.1],
    "Uniform over 4 outcomes [0.25]*4": [0.25, 0.25, 0.25, 0.25],
}

for name, p in examples.items():
    print(f"{name:35s} -> H = {entropy(p):.4f} bits")
```

# 5. Why Entropy Is Maximized by the Uniform Distribution

For a fixed number of possible outcomes, entropy is largest when all outcomes are equally likely.

For example, among all distributions over 4 outcomes, the largest entropy occurs at

$$
P(x)=\frac{1}{4} \quad \text{for all } x
$$

Then

$$
H(X) = - \sum_{i=1}^4 \frac{1}{4}\log_2\left(\frac{1}{4}\right)
= -4 \cdot \frac{1}{4}\cdot (-2) = 2
$$

So entropy is 2 bits.

Why is the uniform distribution the most uncertain?

Because before you observe the outcome, every possibility is equally plausible. You have no bias toward any one of them.

This result is extremely important because it formalizes the intuition that:

> **More spread-out uncertainty means more entropy.**

---

## Visual intuition

If a variable has one dominant outcome, your uncertainty is already partly resolved before observing it.

If all outcomes are equally likely, you cannot predict it well in advance, so uncertainty is maximal.

```
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0.001, 0.999, 500)
H = -(p*np.log2(p) + (1-p)*np.log2(1-p))

plt.figure(figsize=(8,5))
plt.plot(p, H)
plt.xlabel("P(Heads)")
plt.ylabel("Entropy H(X) in bits")
plt.title("Entropy of a Bernoulli Random Variable")
plt.grid(True)
plt.show()
```

# 6. Entropy as Expected Code Length and Compression

One of Shannon's deepest insights is that entropy is linked to **compression**.

Suppose a source emits symbols according to some distribution. If some symbols are much more common than others, we can assign:

- shorter codewords to common symbols
- longer codewords to rare symbols

This is the basic intuition behind efficient coding.

Shannon's source coding theorem says, roughly:

> The entropy $H(X)$ is the fundamental lower limit on the average number of bits needed to encode outcomes from $X$ (in the long run, under ideal coding).

This is why entropy is often interpreted as:

$$
H(X) = \text{minimum achievable average description length}
$$

in bits per symbol, in an idealized setting.

---

## Example intuition

Suppose there are 4 equally likely outcomes.

Then entropy is

$$
H(X)=2 \text{ bits}
$$

This makes sense: 2 bits are enough to distinguish 4 equally likely possibilities.

But if one outcome is extremely common, then the average required number of bits can be less than 2, because we can code the common outcome more efficiently.

# 7. Joint Entropy

Often we care about more than one random variable.

For two discrete random variables $X$ and $Y$, the **joint entropy** is

$$
H(X,Y) = - \sum_{x,y} P(x,y)\log_2 P(x,y)
$$

This measures the uncertainty in the pair $(X,Y)$ together.

---

## Interpretation

Joint entropy answers:

> "How uncertain am I about the combined outcome of both variables?"

If $X$ and $Y$ are independent, then uncertainty in the pair comes from both separately.

If they are strongly dependent, then knowing one may already tell us much about the other, so the joint entropy can be lower than the sum of their separate entropies.

---

## If $X$ and $Y$ are independent

Then

$$
P(x,y)=P(x)P(y)
$$

and therefore

$$
H(X,Y)=H(X)+H(Y)
$$

This is a very important property.

---

## If $X$ and $Y$ are dependent

Then usually

$$
H(X,Y) < H(X) + H(Y)
$$

because there is overlap or redundancy in the uncertainty.

# 8. Conditional Entropy

The **conditional entropy** of $X$ given $Y$ is

$$
H(X \mid Y) = -\sum_{x,y} P(x,y)\log_2 P(x \mid y)
$$

This measures:

> the remaining uncertainty in $X$ after $Y$ is known.

---

## Intuition

Suppose $Y$ tells you something useful about $X$.

Then after learning $Y$, you are less uncertain about $X$.

So $H(X \mid Y)$ should be smaller than $H(X)$.

Indeed, one can show:

$$
H(X \mid Y) \le H(X)
$$

with equality if $Y$ tells you nothing about $X$.

---

## Extreme cases

### Case 1: $Y$ completely determines $X$

Then once you know $Y$, there is no uncertainty left in $X$:

$$
H(X \mid Y)=0
$$

### Case 2: $X$ and $Y$ are independent

Then knowing $Y$ does not help at all, so

$$
H(X \mid Y)=H(X)
$$

---

## Chain rule of entropy

A fundamental identity is

$$
H(X,Y) = H(Y) + H(X \mid Y)
$$

or equally

$$
H(X,Y) = H(X) + H(Y \mid X)
$$

This says:

> Uncertainty in the pair = uncertainty in one variable + remaining uncertainty in the other after the first is known.

# 9. Mutual Information

One of the most important concepts in information theory is **mutual information**.

It measures how much knowing one random variable reduces uncertainty about another.

It is defined as

$$
I(X;Y) = H(X) - H(X \mid Y)
$$

By symmetry, it is also equal to

$$
I(X;Y) = H(Y) - H(Y \mid X)
$$

and also

$$
I(X;Y) = H(X) + H(Y) - H(X,Y)
$$

and also

$$
I(X;Y) = \sum_{x,y} P(x,y)\log_2\frac{P(x,y)}{P(x)P(y)}
$$

---

## Interpretation

Mutual information answers:

> "How much does knowledge of $Y$ tell me about $X$?"

or equally

> "How much uncertainty is shared between them?"

---

## Special cases

### If $X$ and $Y$ are independent

Then

$$
P(x,y)=P(x)P(y)
$$

so

$$
I(X;Y)=0
$$

This makes sense: independent variables tell us nothing about each other.

### If $Y$ completely determines $X$

Then

$$
H(X \mid Y)=0
$$

so

$$
I(X;Y)=H(X)
$$

Knowing $Y$ removes all uncertainty in $X$.

---

## Why mutual information matters in AI

Mutual information appears in:

- feature selection
- representation learning
- self-supervised learning
- contrastive objectives
- information bottleneck ideas
- clustering and dependency analysis

It is a mathematically principled way to talk about how much one variable "contains" about another.

```
import numpy as np

# Example: X and Y are perfectly correlated binary variables
joint = np.array([[0.5, 0.0],
                  [0.0, 0.5]])

px = joint.sum(axis=1)
py = joint.sum(axis=0)

def entropy_from_probs(p):
    p = np.array(p, dtype=float)
    p = p[p > 0]
    return -np.sum(p * np.log2(p))

def mutual_information(joint):
    px = joint.sum(axis=1, keepdims=True)
    py = joint.sum(axis=0, keepdims=True)
    mi = 0.0
    for i in range(joint.shape[0]):
        for j in range(joint.shape[1]):
            if joint[i, j] > 0:
                mi += joint[i, j] * np.log2(joint[i, j] / (px[i, 0] * py[0, j]))
    return mi

print("H(X) =", entropy_from_probs(px))
print("H(Y) =", entropy_from_probs(py))
print("I(X;Y) =", mutual_information(joint))
```

# 10. Cross-Entropy

Cross-entropy is one of the most important quantities in machine learning.

Given two probability distributions $P$ and $Q$ over the same outcomes, the **cross-entropy** of $P$ relative to $Q$ is

$$
H(P,Q) = - \sum_x P(x)\log_2 Q(x)
$$

---

## Interpretation

Suppose:

- $P$ is the true distribution
- $Q$ is the model's predicted distribution

Then cross-entropy measures the average number of bits needed if data actually comes from $P$, but we encode it as though the distribution were $Q$.

If $Q$ matches $P$ well, cross-entropy is low.

If $Q$ is poor, cross-entropy is high.

---

## Relation to entropy

Notice that if $Q=P$, then

$$
H(P,P) = H(P)
$$

So ordinary entropy is a special case of cross-entropy.

---

## Why it appears in classification loss

In classification, the target distribution is often one-hot:

- correct class has probability 1
- all others have probability 0

If the true class is $k$, then cross-entropy becomes

$$
L = -\log \hat{p}_k
$$

where $\hat{p}_k$ is the predicted probability assigned to the correct class.

This is exactly why neural network classification uses cross-entropy / negative log-likelihood.

```
import numpy as np

P = np.array([0.7, 0.2, 0.1])   # imagine the true distribution
Q1 = np.array([0.65, 0.25, 0.10])  # pretty good prediction
Q2 = np.array([0.20, 0.20, 0.60])  # poor prediction

def cross_entropy(P, Q):
    P = np.array(P, dtype=float)
    Q = np.array(Q, dtype=float)
    mask = (P > 0) & (Q > 0)
    return -np.sum(P[mask] * np.log2(Q[mask]))

print("H(P)      =", entropy(P))
print("H(P, Q1)  =", cross_entropy(P, Q1))
print("H(P, Q2)  =", cross_entropy(P, Q2))
```

# 11. KL Divergence

The **Kullback–Leibler divergence** from $Q$ to $P$ is

$$
D_{\mathrm{KL}}(P \parallel Q) = \sum_x P(x)\log_2 \frac{P(x)}{Q(x)}
$$

It measures how different $Q$ is from $P$.

---

## Very important caution

KL divergence is **not a true distance metric**.

Why not?

Because in general:

$$
D_{\mathrm{KL}}(P \parallel Q) \ne D_{\mathrm{KL}}(Q \parallel P)
$$

So it is asymmetric.

---

## Relation to cross-entropy

A very important identity is

$$
D_{\mathrm{KL}}(P \parallel Q) = H(P,Q) - H(P)
$$

This tells us:

- entropy $H(P)$ is the irreducible uncertainty in the true distribution
- cross-entropy $H(P,Q)$ is the coding cost if we use model $Q$
- KL divergence is the **extra cost** due to using the wrong distribution

That is a beautiful and very important interpretation.

---

## Non-negativity

KL divergence is always nonnegative:

$$
D_{\mathrm{KL}}(P \parallel Q) \ge 0
$$

with equality if and only if $P=Q$ almost everywhere.

This is known as **Gibbs' inequality**.

---

## Why KL divergence matters in AI

KL divergence appears everywhere:

- variational autoencoders (VAEs)
- variational inference
- probabilistic modeling
- policy optimization in RL
- distillation and model matching
- distribution alignment
- approximate Bayesian inference

If entropy measures uncertainty, KL divergence measures **mismatch between beliefs/distributions**.

```
import numpy as np

def kl_divergence(P, Q):
    P = np.array(P, dtype=float)
    Q = np.array(Q, dtype=float)
    mask = (P > 0) & (Q > 0)
    return np.sum(P[mask] * np.log2(P[mask] / Q[mask]))

P = np.array([0.7, 0.2, 0.1])
Q = np.array([0.6, 0.3, 0.1])

print("Entropy H(P):", entropy(P))
print("Cross-entropy H(P,Q):", cross_entropy(P, Q))
print("KL divergence D_KL(P || Q):", kl_divergence(P, Q))
print("Check H(P,Q) - H(P):", cross_entropy(P, Q) - entropy(P))
```

# 12. Why Cross-Entropy Is the Standard Classification Loss

This section is especially important for deep learning.

Suppose we have a classification model that outputs probabilities

$$
\hat{\mathbf{y}} = (\hat{y}_1, \hat{y}_2, \dots, \hat{y}_K)
$$

for $K$ classes.

Suppose the true label is class $k$. Then the target distribution is one-hot:

$$
y_i =
\begin{cases}
1 & \text{if } i=k \\
0 & \text{otherwise}
\end{cases}
$$

The cross-entropy loss is

$$
L = -\sum_{i=1}^{K} y_i \log \hat{y}_i
$$

Since all $y_i$ are zero except the correct class, this simplifies to

$$
L = -\log \hat{y}_k
$$

---

## Interpretation

If the model assigns high probability to the correct class, then $\hat{y}_k$ is close to 1, so

$$
-\log \hat{y}_k
$$

is small.

If the model assigns low probability to the correct class, the loss becomes large.

This is exactly what we want.

---

## Why not just use accuracy?

Accuracy is not differentiable and gives no gradient information about *how wrong* a prediction is.

Cross-entropy does.

For example:

- predicting 0.49 for the correct class is bad
- predicting 0.001 for the correct class is much worse

Cross-entropy distinguishes these cases sharply.

That is why it is such a natural fit for gradient-based learning.

```
import numpy as np

correct_class_probs = [0.99, 0.9, 0.7, 0.5, 0.2, 0.01]

print("Cross-entropy / negative log-likelihood for correct class probability:")
for p in correct_class_probs:
    loss = -np.log(p)
    print(f"p_correct = {p:>5} -> loss = {loss:.4f}")
```

# 13. Negative Log-Likelihood and Cross-Entropy

In many machine learning settings, the phrase **negative log-likelihood** (NLL) appears.

This is closely related to cross-entropy.

Suppose our model assigns probability $Q(x)$ to observed data $x$.

The log-likelihood is

$$
\log Q(x)
$$

Maximizing likelihood is equivalent to maximizing log-likelihood, which is equivalent to minimizing negative log-likelihood:

$$
-\log Q(x)
$$

If the targets are one-hot categorical labels, then negative log-likelihood becomes the same as cross-entropy loss.

So these are not separate unrelated ideas. They are deeply connected:

- **maximum likelihood**
- **log-likelihood**
- **cross-entropy**
- **KL divergence**

all live in the same conceptual family.

# 14. Jensen–Shannon Divergence

Because KL divergence is asymmetric, a symmetric alternative is often useful.

The **Jensen–Shannon divergence** between $P$ and $Q$ is

$$
D_{\mathrm{JS}}(P,Q) = \frac{1}{2}D_{\mathrm{KL}}(P \parallel M) + \frac{1}{2}D_{\mathrm{KL}}(Q \parallel M)
$$

where

$$
M = \frac{1}{2}(P+Q)
$$

This quantity is:

- symmetric
- always finite when $P$ and $Q$ are valid distributions
- the square root of JS divergence is a metric

It appears in some generative modeling discussions, including classical GAN literature.

```
import numpy as np

def js_divergence(P, Q):
    P = np.array(P, dtype=float)
    Q = np.array(Q, dtype=float)
    M = 0.5 * (P + Q)
    return 0.5 * kl_divergence(P, M) + 0.5 * kl_divergence(Q, M)

P = np.array([0.7, 0.2, 0.1])
Q = np.array([0.2, 0.2, 0.6])

print("D_JS(P, Q) =", js_divergence(P, Q))
print("D_JS(Q, P) =", js_divergence(Q, P))
```

# 15. Information Gain in Decision Trees

Information theory directly powers decision tree learning.

Suppose we want to classify data. At each node, the tree considers candidate splits.

A good split should reduce uncertainty about the label.

That reduction is measured by **information gain**:

$$
\text{Information Gain} = H(Y) - H(Y \mid \text{split})
$$

Here:

- $H(Y)$ = entropy of labels before the split
- $H(Y \mid \text{split})$ = remaining entropy after partitioning by the split

A split is good if it makes the child nodes purer, meaning labels become more predictable.

---

## Why entropy is appropriate here

If a node contains a 50-50 mix of two classes, uncertainty is high.

If a node contains almost all one class, uncertainty is low.

So entropy gives a principled way to measure node impurity.

This is one of the most concrete and classical AI applications of information theory.

```
import numpy as np

def binary_entropy(p):
    if p == 0 or p == 1:
        return 0.0
    return -(p*np.log2(p) + (1-p)*np.log2(1-p))

# Example:
# Parent node: 10 samples, 5 positive / 5 negative
parent_entropy = binary_entropy(0.5)

# Split into:
# Left child: 4 positive / 1 negative  => p=0.8
# Right child: 1 positive / 4 negative => p=0.2
left_entropy = binary_entropy(0.8)
right_entropy = binary_entropy(0.2)

weighted_child_entropy = 0.5 * left_entropy + 0.5 * right_entropy
information_gain = parent_entropy - weighted_child_entropy

print("Parent entropy:", parent_entropy)
print("Weighted child entropy:", weighted_child_entropy)
print("Information gain:", information_gain)
```

# 16. Entropy in Reinforcement Learning

Entropy also appears in reinforcement learning (RL), especially in **policy optimization**.

Suppose a policy outputs action probabilities $\pi(a \mid s)$ in a state $s$.

The policy entropy is

$$
H(\pi(\cdot \mid s)) = - \sum_a \pi(a \mid s)\log \pi(a \mid s)
$$

Why might we care?

Because if the policy becomes too certain too early, it may stop exploring.

So many RL methods add an **entropy bonus** to encourage exploration.

The objective may look like:

$$
J = \mathbb{E}[\text{return}] + \beta H(\pi(\cdot \mid s))
$$

where $\beta$ controls how much exploration is encouraged.

---

## Intuition

- high entropy policy = more exploratory, less certain
- low entropy policy = more deterministic, more exploitative

Information theory here gives a clean mathematical handle on the exploration–exploitation tradeoff.

# 17. Differential Entropy

So far we have discussed discrete random variables.

For continuous random variables, the analogue is **differential entropy**:

$$
h(X) = - \int p(x)\log p(x)\, dx
$$

This looks similar to discrete entropy, but there are important caveats.

---

## Important caution

Differential entropy behaves differently from discrete entropy.

In particular:

- it can be negative
- it is not directly interpretable as expected code length in the same simple way
- only relative quantities like KL divergence often remain more robustly meaningful

---

## Example: Gaussian differential entropy

For a one-dimensional Gaussian distribution $X \sim \mathcal{N}(\mu, \sigma^2)$,

$$
h(X) = \frac{1}{2}\log(2\pi e \sigma^2)
$$

This shows that broader Gaussians (larger variance) have larger differential entropy.

That matches the intuition that more spread means more uncertainty.

Differential entropy becomes important in continuous probabilistic modeling, variational inference, and statistics.

```
import numpy as np

def gaussian_differential_entropy(sigma):
    return 0.5 * np.log(2 * np.pi * np.e * sigma**2)

sigmas = [0.1, 0.5, 1.0, 2.0]
for s in sigmas:
    print(f"sigma = {s:.1f} -> h(X) = {gaussian_differential_entropy(s):.4f} nats")
```

# 18. Channel Capacity (High-Level Introduction)

Information theory originally arose from communication problems.

Suppose we have:

- a sender
- a noisy channel
- a receiver

The sender wants to transmit information reliably through noise.

Shannon asked:

> What is the maximum rate at which information can be transmitted with arbitrarily low error?

The answer is the **channel capacity**:

$$
C = \max_{P(X)} I(X;Y)
$$

where:

- $X$ is the channel input
- $Y$ is the channel output
- $I(X;Y)$ is the mutual information between input and output

So channel capacity is the maximum mutual information achievable by choosing the best input distribution.

---

## Why this matters conceptually for AI

Even if you are not designing communication systems, this idea is profound:

> Information transfer is fundamentally limited by noise and representation.

That intuition carries over into AI when we think about:

- compression
- representation bottlenecks
- noisy observations
- latent variables
- communication between layers/modules

# 19. Rate–Distortion Theory (High-Level Intuition)

Another major branch of information theory asks:

> If we are allowed to compress data with some loss, what is the minimum rate needed to achieve a desired fidelity?

This is the domain of **rate–distortion theory**.

At a high level:

- **rate** = how many bits we use
- **distortion** = how much error/loss we allow

The rate–distortion function gives the best tradeoff.

Why is this relevant today?

Because many modern ML systems implicitly live in a rate–distortion world:

- autoencoders
- lossy compression
- representation bottlenecks
- generative latent spaces

You may not see the exact Shannon-style formulas in every deep learning paper, but the intuition is deeply connected.

# 20. Mutual Information in Representation Learning

Modern representation learning often wants a latent representation $Z$ that captures useful information about input $X$ and perhaps about labels or future views.

Several ideas can be phrased in mutual-information language:

- maximize $I(Z;Y)$ to keep task-relevant information
- control or reduce $I(Z;X)$ to compress irrelevant details
- compare different views of the same underlying object

Contrastive learning methods often have interpretations related to mutual information estimation or lower bounds on mutual information.

This area can become mathematically subtle, but the core idea is simple:

> A good representation should preserve useful information while discarding useless noise.

That sentence is very information-theoretic.

# 21. Information Bottleneck Intuition

The **information bottleneck** perspective asks for a representation $Z$ that:

- compresses the input $X$
- preserves information relevant to the target $Y$

This is often expressed as a tradeoff involving

$$
I(Z;X) \quad \text{and} \quad I(Z;Y)
$$

A common qualitative statement is:

- keep $I(Z;Y)$ high
- keep $I(Z;X)$ low

That means:

- retain task-relevant information
- discard unnecessary detail

This is a very appealing lens for thinking about deep learning representations.

---

## Why it is conceptually powerful

It provides a formal way of expressing something we often say informally:

> "A good representation should be compact but informative."

That sentence almost *is* the information bottleneck idea in plain English.

---

## Caveat

The exact role of information bottleneck in practical deep learning is subtle and debated in some contexts. But as a conceptual framework, it is extremely useful.

# 22. KL Divergence in Variational Autoencoders (VAEs)

Since VAEs are a major AI application of information-theoretic quantities, let us state the key idea carefully.

A VAE learns:

- an encoder that maps input $x$ to a distribution over latent variable $z$
- a decoder that reconstructs $x$ from $z$

The training objective includes two parts:

$$
\mathcal{L} = \text{reconstruction term} + D_{\mathrm{KL}}(q(z \mid x)\parallel p(z))
$$

More commonly, one maximizes the ELBO:

$$
\mathrm{ELBO}
=
\mathbb{E}_{q(z \mid x)}[\log p(x \mid z)]
-
D_{\mathrm{KL}}(q(z \mid x)\parallel p(z))
$$

---

## Interpretation of the KL term

The encoder produces a learned latent distribution $q(z \mid x)$.

The prior is usually a simple distribution like $p(z)=\mathcal{N}(0,I)$.

The KL term penalizes the encoder if its latent distribution drifts too far from the prior.

This encourages:

- smooth latent space
- regularization
- meaningful generative sampling

So KL divergence here measures how much the encoder's posterior-like distribution differs from the desired prior.

This is one of the most famous appearances of KL divergence in deep learning.

# 23. Why KL Divergence Is Asymmetric — Intuition

The asymmetry of KL divergence is often confusing.

Let us unpack it.

Recall:

$$
D_{\mathrm{KL}}(P \parallel Q) = \sum_x P(x)\log \frac{P(x)}{Q(x)}
$$

This is an expectation under $P$.

That means it weights errors according to what is likely under the true distribution $P$.

So if $Q$ badly underestimates an event that $P$ considers important, the penalty can be large.

But if $Q$ puts mass in places where $P$ has almost none, the penalty behaves differently.

This is why the direction matters.

---

## Practical intuition

- $D_{\mathrm{KL}}(P \parallel Q)$ asks: how costly is it to use $Q$ when reality is $P$?
- $D_{\mathrm{KL}}(Q \parallel P)$ asks a different question: how costly is it to think the world is $Q$ when compared against $P$?

They are not the same question, so they are not the same number.

# 24. Common Confusions and Important Clarifications

This section is especially important.

## Confusion 1: "Information" means semantic meaning

Not in Shannon information theory.

A random string of noise may contain lots of Shannon information if it is unpredictable, even if it has little meaning to a human.

Shannon's theory is about **uncertainty and coding**, not meaning.

---

## Confusion 2: Higher entropy always means "better"

Not necessarily.

High entropy means greater uncertainty or spread.

Whether that is desirable depends on context.

Examples:

- In compression, higher entropy means data is harder to compress.
- In exploration for RL, higher policy entropy can be useful.
- In prediction, lower uncertainty may be better if the model is correct.

---

## Confusion 3: KL divergence is a distance

It is not a metric because:
- it is asymmetric
- it does not satisfy the triangle inequality

---

## Confusion 4: Cross-entropy and KL divergence are completely different

They are closely related:

$$
D_{\mathrm{KL}}(P \parallel Q)=H(P,Q)-H(P)
$$

So minimizing cross-entropy with respect to $Q$ is equivalent to minimizing KL divergence when $P$ is fixed.

---

## Confusion 5: Entropy is only a communication-theory topic

No. It is deeply connected to machine learning, optimization, uncertainty, Bayesian inference, and deep learning objectives.

# 25. Worked Numerical Example by Hand

Let a random variable $X$ take 3 values with probabilities

$$
P(X=a)=0.5,\quad P(X=b)=0.25,\quad P(X=c)=0.25
$$

Let us compute the entropy carefully.

By definition,

$$
H(X) = - \sum_x P(x)\log_2 P(x)
$$

So

$$
H(X)
=
- \left[
0.5\log_2(0.5)
+
0.25\log_2(0.25)
+
0.25\log_2(0.25)
\right]
$$

Now use:

$$
\log_2(0.5)=-1,\qquad \log_2(0.25)=-2
$$

Then

$$
H(X)
=
- \left[
0.5(-1) + 0.25(-2) + 0.25(-2)
\right]
$$

$$
=
- \left[
-0.5 - 0.5 - 0.5
\right]
$$

$$
= 1.5 \text{ bits}
$$

So on average, observing this variable gives 1.5 bits of information.

That is a nice example because it is more uncertain than a biased coin, but less uncertain than four equally likely outcomes (which would give 2 bits).

```
import numpy as np

P = np.array([0.5, 0.25, 0.25])
print("Entropy of [0.5, 0.25, 0.25] =", entropy(P), "bits")
```

I(x) = -\log_2 P(x)
$$

As $P(x)$ becomes smaller, the information rises sharply.

This is why very rare events are mathematically "surprising".

# 27. A Coding Interpretation Example

Suppose a source emits 4 symbols with probabilities:

$$
P(A)=0.5,\quad P(B)=0.25,\quad P(C)=0.125,\quad P(D)=0.125
$$

The self-information values are:

$$
I(A)=1,\quad I(B)=2,\quad I(C)=3,\quad I(D)=3
$$

This suggests a coding scheme where:

- common symbols get short codes
- rare symbols get longer codes

For example:

- $A \rightarrow 0$
- $B \rightarrow 10$
- $C \rightarrow 110$
- $D \rightarrow 111$

This is not a full derivation of Huffman coding, but it shows the principle:

> Efficient coding aligns code length with information content.

That is one of the grand ideas behind information theory.

# 28. Entropy, Likelihood, and Model Training

A very important machine learning viewpoint is this:

When we train a probabilistic model, we want the model to assign high probability to observed data.

High probability means high log-probability.

So we maximize

$$
\log p_\theta(x)
$$

or equivalently minimize

$$
-\log p_\theta(x)
$$

When averaged over the data distribution, this connects directly to cross-entropy and KL divergence.

This is why information theory is not some side topic. It is embedded directly in modern model training.

# 29. Information Theory and Language Models

Language modeling is a beautiful real-world example.

A language model predicts the next token distribution:

$$
p_\theta(x_t \mid x_{<t})
$$

Training typically minimizes the negative log-likelihood of the true next token, which is equivalent to cross-entropy loss.

So when a language model is trained, information-theoretic quantities are being optimized at every step.

Metrics like **perplexity** are also directly derived from cross-entropy.

If the average negative log-likelihood is low, perplexity is low, meaning the model is less "surprised" by the true sequence.

# 30. Perplexity

Perplexity is a standard metric in language modeling.

If the cross-entropy (in bits) is $H$, then perplexity is

$$
\mathrm{Perplexity} = 2^H
$$

If cross-entropy is measured in nats, then perplexity is often written as

$$
\mathrm{Perplexity} = e^H
$$

depending on the log base used.

---

## Interpretation

Perplexity can be thought of as the model's effective branching factor.

A lower perplexity means the model is less uncertain about the next token.

This is yet another direct bridge between information theory and practical AI.

```
import numpy as np

cross_entropy_bits = [1.0, 2.0, 3.0, 4.0]
for h in cross_entropy_bits:
    ppl = 2**h
    print(f"Cross-entropy = {h:.1f} bits -> Perplexity = {ppl:.2f}")
```

# 31. Conditional Entropy and Label Noise

Conditional entropy gives a useful way to think about noisy supervision.

Suppose $Y$ is the label and $X$ is the input.

If $H(Y \mid X)$ is high, then even after seeing the input, there remains uncertainty in the label. This may happen due to:

- label noise
- ambiguity
- insufficient features
- overlapping classes

If $H(Y \mid X)$ is low, then the input strongly determines the label.

So conditional entropy is a good conceptual tool for thinking about the difficulty of prediction tasks.

# 32. Mutual Information and Feature Selection

Suppose we have many candidate input features $X_1, X_2, \dots$ and a label $Y$.

A natural question is:

> Which feature tells us the most about the label?

One principled answer is: choose features with high mutual information $I(X_i;Y)$.

That means learning the feature reduces uncertainty about the target.

This does not solve all feature-selection problems, but it gives a rigorous criterion rooted in information theory.

# 33. A Compact Summary of the Core Quantities

It helps to see the major information-theoretic quantities side by side.

## Self-information

$$
I(x) = -\log P(x)
$$

Information in one event.

## Entropy

$$
H(X) = -\sum_x P(x)\log P(x)
$$

Average uncertainty of one variable.

## Joint entropy

$$
H(X,Y) = -\sum_{x,y} P(x,y)\log P(x,y)
$$

Uncertainty of two variables together.

## Conditional entropy

$$
H(X \mid Y) = -\sum_{x,y} P(x,y)\log P(x \mid y)
$$

Remaining uncertainty in $X$ after knowing $Y$.

## Mutual information

$$
I(X;Y)=H(X)-H(X \mid Y)
$$

Reduction in uncertainty in one variable due to knowing the other.

## Cross-entropy

$$
H(P,Q) = -\sum_x P(x)\log Q(x)
$$

Coding cost if truth is $P$ but model is $Q$.

## KL divergence

$$
D_{\mathrm{KL}}(P \parallel Q)=\sum_x P(x)\log\frac{P(x)}{Q(x)}
$$

Extra coding cost / mismatch between $P$ and $Q$.

```
# Quick sandbox cell:
# Edit these probabilities and rerun to build intuition.

import numpy as np

P = np.array([0.5, 0.3, 0.2])
Q = np.array([0.4, 0.4, 0.2])

def entropy(p):
    p = np.array(p, dtype=float)
    p = p[p > 0]
    return -np.sum(p * np.log2(p))

def cross_entropy(P, Q):
    P = np.array(P, dtype=float)
    Q = np.array(Q, dtype=float)
    mask = (P > 0) & (Q > 0)
    return -np.sum(P[mask] * np.log2(Q[mask]))

def kl_divergence(P, Q):
    P = np.array(P, dtype=float)
    Q = np.array(Q, dtype=float)
    mask = (P > 0) & (Q > 0)
    return np.sum(P[mask] * np.log2(P[mask] / Q[mask]))

print("P =", P)
print("Q =", Q)
print("H(P) =", entropy(P))
print("H(P,Q) =", cross_entropy(P, Q))
print("D_KL(P || Q) =", kl_divergence(P, Q))
```
