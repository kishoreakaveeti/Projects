## Core idea in one sentence

A matrix usually changes both the **length** and the **direction** of a vector, but an **eigenvector** is a special vector whose direction remains unchanged under the transformation; the **eigenvalue** tells how much that vector is scaled.

# 1. First intuition: what problem are eigenvalues and eigenvectors solving?

Suppose a matrix $A$ represents a linear transformation.

A linear transformation can do things like:

- stretch
- shrink
- flip
- rotate
- shear

If we apply the matrix $A$ to a vector $x$, we get a new vector:

$$
Ax
$$

For **most** vectors, the output points in a different direction from the input.

But some vectors are special.

There may exist certain nonzero vectors $v$ such that after applying the transformation, the output remains along the **same line** as the original vector:

$$
Av = \lambda v
$$

This means:

- the direction is preserved
- only the magnitude changes
- if $\lambda < 0$, the direction flips but still stays on the same line

These special vectors are called **eigenvectors**, and the corresponding scalars $\lambda$ are called **eigenvalues**.

---

## Very plain-language version

If a matrix is a machine that deforms space, then:

- **eigenvectors** are the directions that the machine treats in a "pure" way
- **eigenvalues** tell how strongly the machine stretches or shrinks along those directions

---

## Why this matters

This is important because a complicated transformation often becomes much easier to understand if we can identify its special directions.

That is exactly what eigenvalues and eigenvectors do.

They reveal the matrix's **natural directions** and **natural scaling behavior**.



# 2. Formal definition

For a square matrix $A \in \mathbb{R}^{n \times n}$, a nonzero vector $v$ is called an eigenvector of $A$ if there exists a scalar $\lambda$ such that

$$
Av = \lambda v
$$

where:

- $A$ is the matrix
- $v$ is the eigenvector
- $\lambda$ is the eigenvalue

---

## Important details

### $A$ is a matrix
Yes — $A$ is the transformation matrix.

### $\lambda$ is a scalar
Yes — $\lambda$ is just a number.

### $v$ must be nonzero
We exclude the zero vector because:

$$
A0 = \lambda 0 = 0
$$

for any $\lambda$, so the zero vector gives no useful information.

---

## What the equation means

The left side:

$$
Av
$$

means "apply the transformation $A$ to the vector $v$."

The right side:

$$
\lambda v
$$

means "take the same vector $v$ and scale it by $\lambda$."

So the equation says:

> applying the transformation to $v$ has exactly the same effect as simply scaling $v$

That is what makes $v$ special.



# 3. Geometric interpretation

This is the most important intuition.

A matrix can move vectors around in complicated ways. But eigenvectors are the vectors whose direction does not get "mixed."

For an eigenvector $v$:

- if $\lambda > 1$, it gets stretched
- if $0 < \lambda < 1$, it gets shrunk
- if $\lambda = 1$, it stays the same
- if $\lambda = 0$, it collapses to zero
- if $\lambda < 0$, it flips direction and gets scaled by $|\lambda|$

---

## Important clarification

When we say "direction stays the same," we really mean:

- the output still lies on the same line through the origin

If $\lambda$ is negative, the vector points the other way, but it is still on that same line.

So eigenvectors represent **special invariant lines** of the transformation.

---

## Another key point

If $v$ is an eigenvector, then any nonzero scalar multiple of $v$ is also an eigenvector.

That is because if:

$$
Av = \lambda v
$$

then for any scalar $c \neq 0$,

$$
A(cv) = cAv = c(\lambda v) = \lambda (cv)
$$

So eigenvectors are really about **directions/lines**, not just one specific arrow.


```python

import numpy as np

A = np.array([[2, 0],
              [0, 3]])

v1 = np.array([1, 0])
v2 = np.array([0, 1])
u  = np.array([1, 1])

print("A =\n", A)
print("\nA @ v1 =", A @ v1)
print("A @ v2 =", A @ v2)
print("A @ u  =", A @ u)
```


# 4. First concrete example

Consider

$$
A =
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix}
$$

This matrix stretches the $x$-coordinate by 2 and the $y$-coordinate by 3.

Now take

$$
v_1 =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

Then

$$
Av_1 =
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}
=
\begin{bmatrix}
2 \\
0
\end{bmatrix}
=
2
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

So $v_1$ is an eigenvector with eigenvalue $2$.

Now take

$$
v_2 =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

Then

$$
Av_2 =
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}
=
\begin{bmatrix}
0 \\
3
\end{bmatrix}
=
3
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

So $v_2$ is an eigenvector with eigenvalue $3$.

---

## Why $\begin{bmatrix}1 \\ 1\end{bmatrix}$ is not an eigenvector

Let

$$
u =
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

Then

$$
Au =
\begin{bmatrix}
2 \\
3
\end{bmatrix}
$$

This is **not** a scalar multiple of $u$, because the coordinates are not scaled by the same factor.

So $u$ is not an eigenvector.

---

## Interpretation

This transformation has two natural directions:

- along the $x$-axis, scale by 2
- along the $y$-axis, scale by 3

Those are the eigendirections.


```python

import matplotlib.pyplot as plt

A = np.array([[2, 0],
              [0, 3]])

vectors = {
    "v1 = [1,0]": np.array([1, 0]),
    "v2 = [0,1]": np.array([0, 1]),
    "u = [1,1]": np.array([1, 1]),
}

plt.figure(figsize=(7, 7))
plt.axhline(0)
plt.axvline(0)

for label, vec in vectors.items():
    transformed = A @ vec
    plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1)
    plt.quiver(0, 0, transformed[0], transformed[1], angles='xy', scale_units='xy', scale=1)
    plt.text(vec[0] + 0.05, vec[1] + 0.05, label)
    plt.text(transformed[0] + 0.05, transformed[1] + 0.05, "A@" + label.split(" = ")[0])

plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Original vectors and transformed vectors under A")
plt.show()
```


# 5. Why the characteristic equation appears

We start from the defining equation:

$$
Av = \lambda v
$$

Bring everything to one side:

$$
Av - \lambda v = 0
$$

Now factor out $v$:

$$
(A - \lambda I)v = 0
$$

Here $I$ is the identity matrix.

---

## Why introduce $I$?

Because $\lambda$ is a scalar, while $A$ is a matrix. To subtract $\lambda$ from the matrix in a meaningful way, we write:

$$
\lambda I
$$

So:

$$
A - \lambda I
$$

means subtract $\lambda$ from the diagonal entries of $A$.

---

## Why determinant must be zero

The equation

$$
(A - \lambda I)v = 0
$$

is a homogeneous linear system.

A homogeneous system always has the trivial solution $v = 0$, but we want a **nonzero** solution.

A nonzero solution exists only if the matrix $A - \lambda I$ is singular.

A matrix is singular exactly when its determinant is zero.

So we require:

$$
\det(A - \lambda I) = 0
$$

This equation is called the **characteristic equation**, and its roots are the eigenvalues.

---

## Summary of the logic

1. Start with:

$$
Av = \lambda v
$$

2. Rearrange:

$$
(A - \lambda I)v = 0
$$

3. For nonzero $v$, require singularity:

$$
\det(A - \lambda I) = 0
$$

4. Solve for $\lambda$

5. Plug each $\lambda$ back in and solve for $v$

This is the standard procedure for finding eigenvalues and eigenvectors.



# 6. Worked example step by step

Take

$$
A =
\begin{bmatrix}
3 & 1 \\
0 & 2
\end{bmatrix}
$$

We want to find its eigenvalues and eigenvectors.

---

## Step 1: Form $A - \lambda I$

$$
A - \lambda I =
\begin{bmatrix}
3-\lambda & 1 \\
0 & 2-\lambda
\end{bmatrix}
$$

---

## Step 2: Compute the determinant

$$
\det(A - \lambda I)
=
\begin{vmatrix}
3-\lambda & 1 \\
0 & 2-\lambda
\end{vmatrix}
=
(3-\lambda)(2-\lambda)
$$

Set this equal to zero:

$$
(3-\lambda)(2-\lambda)=0
$$

So the eigenvalues are:

$$
\lambda_1 = 3, \qquad \lambda_2 = 2
$$

---

## Step 3: Find the eigenvector for $\lambda = 3$

Solve:

$$
(A - 3I)v = 0
$$

$$
\begin{bmatrix}
0 & 1 \\
0 & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

This gives:

$$
y = 0
$$

So $x$ can be any nonzero value.

One eigenvector is:

$$
v_1 =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

---

## Step 4: Find the eigenvector for $\lambda = 2$

Solve:

$$
(A - 2I)v = 0
$$

$$
\begin{bmatrix}
1 & 1 \\
0 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

This gives:

$$
x + y = 0 \Rightarrow y = -x
$$

So one eigenvector is:

$$
v_2 =
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
$$

---

## Interpretation

The matrix has two eigendirections:

- $\begin{bmatrix}1 \\ 0\end{bmatrix}$, scaled by 3
- $\begin{bmatrix}1 \\ -1\end{bmatrix}$, scaled by 2

So even though the matrix is not diagonal, it still has its own preferred directions.


```python

A = np.array([[3, 1],
              [0, 2]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print("A =\n", A)
print("\nEigenvalues:")
print(eigenvalues)
print("\nEigenvectors (each column is an eigenvector):")
print(eigenvectors)

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]
    left = A @ v
    right = lam * v
    print(f"\nChecking eigenpair {i+1}:")
    print("lambda =", lam)
    print("v =", v)
    print("A @ v =", left)
    print("lambda * v =", right)
```


# 7. Meanings of different eigenvalues

The sign and magnitude of $\lambda$ matter.

## Case 1: $\lambda > 1$

The eigenvector is stretched.

Example:

$$
Av = 4v
$$

The vector becomes 4 times longer.

---

## Case 2: $0 < \lambda < 1$

The eigenvector shrinks.

Example:

$$
Av = 0.2v
$$

Same direction, shorter magnitude.

---

## Case 3: $\lambda = 1$

No change in length.

Example:

$$
Av = v
$$

The transformation leaves that vector unchanged.

---

## Case 4: $\lambda = 0$

That direction gets crushed to zero.

Example:

$$
Av = 0
$$

This means that along that direction, information is completely removed.

---

## Case 5: $\lambda < 0$

The vector flips direction and gets scaled.

Example:

$$
Av = -3v
$$

The vector points the opposite way and becomes 3 times longer in magnitude.

---

## Why $\lambda = 0$ is especially important

If a matrix has an eigenvalue 0, then the matrix is not invertible.

That is because some nonzero vector gets mapped to zero, which means information is lost.

This idea matters a lot in:

- rank
- null space
- degeneracy
- compression
- dimensional collapse



# 8. Example with a zero eigenvalue

Take

$$
A =
\begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix}
$$

Find eigenvalues:

$$
A - \lambda I =
\begin{bmatrix}
1-\lambda & 1 \\
1 & 1-\lambda
\end{bmatrix}
$$

$$
\det(A - \lambda I) = (1-\lambda)^2 - 1
$$

$$
= 1 - 2\lambda + \lambda^2 - 1
= \lambda^2 - 2\lambda
= \lambda(\lambda - 2)
$$

So the eigenvalues are:

$$
\lambda = 0, \qquad \lambda = 2
$$

Now check the vector

$$
v =
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
$$

Then

$$
Av =
\begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix}
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

So this is an eigenvector with eigenvalue 0.

---

## Interpretation

The direction $\begin{bmatrix}1 \\ -1\end{bmatrix}$ gets completely collapsed.

The matrix preserves some directions and destroys others.

This idea is deeply related to dimensionality reduction and loss of information.


```python

A = np.array([[1, 1],
              [1, 1]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors (columns):\n", eigenvectors)

v = np.array([1, -1], dtype=float)
print("\nA @ [1, -1] =", A @ v)
```


# 9. Example of a matrix with no real eigenvectors

Consider the 2D rotation matrix for a $90^\circ$ rotation:

$$
R =
\begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
$$

This rotates every vector by $90^\circ$.

Can any nonzero real vector remain on the same line after a $90^\circ$ rotation?

No.

So there are no real eigenvectors.

---

## Why this is conceptually important

It shows that not every real matrix has real eigenvectors.

Eigenvectors exist only when the transformation has directions that remain invariant under the transformation.

A pure rotation in 2D does not preserve any real direction.

If we allow complex numbers, then eigenvalues exist. But in real geometry, there are no real eigendirections here.


```python

R = np.array([[0, -1],
              [1,  0]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(R)

print("R =\n", R)
print("\nEigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```


# 10. Diagonalization: why eigenvectors are powerful

If a matrix has enough linearly independent eigenvectors, then we can write it as:

$$
A = P \Lambda P^{-1}
$$

where:

- $P$ is the matrix whose columns are eigenvectors
- $\Lambda$ is a diagonal matrix whose diagonal entries are eigenvalues

---

## Why this is valuable

A diagonal matrix is much easier to understand than a general matrix.

Instead of mixing coordinates, a diagonal matrix simply scales each coordinate independently.

That means:

$$
A^k = P \Lambda^k P^{-1}
$$

and $\Lambda^k$ is very easy to compute because:

$$
\Lambda^k =
\begin{bmatrix}
\lambda_1^k & 0 & \cdots & 0 \\
0 & \lambda_2^k & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & \lambda_n^k
\end{bmatrix}
$$

So repeated applications of $A$ become much easier to analyze.

---

## Deep intuition

Eigenvectors provide a coordinate system in which the transformation becomes as simple as possible.

That is one of the deepest reasons eigenvectors matter.

They reveal the matrix's "natural basis."


```python

A = np.array([[2, 1],
              [1, 2]], dtype=float)

eigenvalues, P = np.linalg.eig(A)
Lambda = np.diag(eigenvalues)
Pinv = np.linalg.inv(P)

reconstructed = P @ Lambda @ Pinv

print("A =\n", A)
print("\nP =\n", P)
print("\nLambda =\n", Lambda)
print("\nP^-1 =\n", Pinv)
print("\nP @ Lambda @ P^-1 =\n", reconstructed)
print("\nAll close to A?", np.allclose(A, reconstructed))
```


# 11. Repeated multiplication and dynamical systems intuition

Suppose we have a system that evolves as:

$$
x_{t+1} = A x_t
$$

Then after $t$ steps:

$$
x_t = A^t x_0
$$

If $A$ is diagonalizable, then:

$$
x_t = P \Lambda^t P^{-1} x_0
$$

This means each eigen-component evolves according to powers of the corresponding eigenvalue:

$$
\lambda_1^t,\ \lambda_2^t,\ \ldots
$$

---

## Why this matters

If $|\lambda| > 1$, that component grows.

If $|\lambda| < 1$, that component decays.

If one eigenvalue has the largest magnitude, then over many steps that direction often dominates.

This idea appears in:

- dynamical systems
- iterative algorithms
- recurrent neural networks
- gradient flow intuition
- Markov processes


```python

A = np.array([[2.0, 0.0],
              [0.0, 0.5]])

x = np.array([1.0, 1.0])

trajectory = [x.copy()]
for _ in range(8):
    x = A @ x
    trajectory.append(x.copy())

trajectory = np.array(trajectory)
trajectory
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(trajectory[:, 0], marker='o', label='x-component')
plt.plot(trajectory[:, 1], marker='o', label='y-component')
plt.title("Repeated multiplication by A")
plt.xlabel("Step")
plt.ylabel("Component value")
plt.legend()
plt.show()
```


# 12. Related topic: symmetric matrices

Symmetric matrices are especially important.

A matrix is symmetric if:

$$
A = A^T
$$

For symmetric matrices, some beautiful facts hold:

1. All eigenvalues are real.
2. Eigenvectors corresponding to different eigenvalues are orthogonal.
3. The matrix can be diagonalized as:

$$
A = Q \Lambda Q^T
$$

where $Q$ is an orthogonal matrix.

---

## Why this matters

This makes the geometry extremely clean.

Symmetric matrices appear everywhere in AI and ML, especially in:

- covariance matrices
- Hessian approximations
- quadratic forms
- kernel methods
- similarity structures

This is also why PCA is so elegant: the covariance matrix is symmetric.



# 13. Related topic: covariance matrices and why PCA uses eigenvectors

Suppose we have a dataset with multiple features.

After centering the data, the covariance matrix tells us:

- how much each feature varies
- how features vary together

If the covariance matrix is $C$, then PCA finds eigenvectors and eigenvalues of $C$.

---

## Meaning in PCA

- **Eigenvectors** = principal directions in feature space
- **Eigenvalues** = amount of variance captured along those directions

The principal component with the largest eigenvalue is the direction along which the data varies the most.

The second principal component is the next most important orthogonal direction, and so on.

---

## Why that is useful

This allows us to:

- reduce dimensionality
- compress data
- remove noise
- visualize high-dimensional data
- extract informative features

That is why eigenvalues and eigenvectors are foundational in machine learning.



# 14. PCA-style worked example by hand

Suppose the covariance matrix is:

$$
C =
\begin{bmatrix}
4 & 2 \\
2 & 4
\end{bmatrix}
$$

Find eigenvalues:

$$
\det(C - \lambda I)
=
\begin{vmatrix}
4-\lambda & 2 \\
2 & 4-\lambda
\end{vmatrix}
=
(4-\lambda)^2 - 4
$$

$$
= 16 - 8\lambda + \lambda^2 - 4
= \lambda^2 - 8\lambda + 12
= (\lambda - 6)(\lambda - 2)
$$

So the eigenvalues are:

$$
6,\qquad 2
$$

Now for $\lambda = 6$:

$$
(C - 6I)v =
\begin{bmatrix}
-2 & 2 \\
2 & -2
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
= 0
$$

This gives:

$$
y = x
$$

So one eigenvector is:

$$
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

Now for $\lambda = 2$:

$$
(C - 2I)v =
\begin{bmatrix}
2 & 2 \\
2 & 2
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
= 0
$$

This gives:

$$
x + y = 0 \Rightarrow y = -x
$$

So one eigenvector is:

$$
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
$$

---

## Interpretation

The data varies most strongly along the direction:

$$
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

because its eigenvalue is 6.

It varies less strongly along:

$$
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
$$

because its eigenvalue is 2.

If we want a 1D summary of the data, we keep the first direction.


```python

np.random.seed(0)

# Create synthetic 2D data elongated roughly along a diagonal direction
mean = np.array([0, 0])
cov = np.array([[4, 2],
                [2, 4]])

X = np.random.multivariate_normal(mean, cov, size=300)

X_centered = X - X.mean(axis=0)
C = np.cov(X_centered.T)

eigenvalues, eigenvectors = np.linalg.eig(C)

# Sort by descending eigenvalue
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

print("Estimated covariance matrix:\n", C)
print("\nEigenvalues:", eigenvalues)
print("\nEigenvectors (columns):\n", eigenvectors)
```

```python

plt.figure(figsize=(7, 7))
plt.scatter(X_centered[:, 0], X_centered[:, 1], alpha=0.35)

origin = np.array([0, 0])
for i in range(2):
    vec = eigenvectors[:, i] * np.sqrt(eigenvalues[i]) * 3
    plt.quiver(*origin, vec[0], vec[1], angles='xy', scale_units='xy', scale=1)
    plt.quiver(*origin, -vec[0], -vec[1], angles='xy', scale_units='xy', scale=1)

plt.axhline(0)
plt.axvline(0)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Centered data with principal directions")
plt.show()
```


# 15. AI application: PCA and dimensionality reduction

PCA is one of the most famous applications of eigenvalues and eigenvectors in AI and machine learning.

## What PCA is doing conceptually

Suppose your data has many features, but the data actually lies near a lower-dimensional structure.

For example:

- image data may live near a lower-dimensional manifold
- tabular features may be strongly correlated
- embeddings may contain redundant directions

PCA finds the most informative directions in which the data varies.

These directions are given by the eigenvectors of the covariance matrix.

The amount of signal/variance in each direction is given by the eigenvalues.

---

## Why this matters in practice

PCA can help with:

- compression
- visualization
- denoising
- feature extraction
- preprocessing before training downstream models

Even when PCA is not the final production tool, the underlying idea — finding dominant directions in data — is central to ML.



# 16. AI application: spectral clustering and graph methods

Eigenvectors also appear in graph-based learning methods.

If data points are represented as nodes in a graph, then one can build matrices such as:

- adjacency matrices
- degree matrices
- graph Laplacians

The eigenvectors of graph Laplacians reveal important structure.

They can help identify:

- clusters
- community boundaries
- smooth directions on the graph
- connected components

This is the basis of **spectral clustering** and related methods.

---

## Intuition

A graph Laplacian contains information about how nodes are connected.

Its eigenvectors reveal the large-scale organization of the graph.

So once again, eigenvectors are giving us the "natural structure" of the system.



# 17. AI application: optimization, Hessians, and curvature

In deep learning, optimization often depends on the geometry of the loss surface.

A key object here is the **Hessian** matrix, which contains second derivatives.

The eigenvectors of the Hessian indicate principal curvature directions.

The eigenvalues indicate the strength and sign of curvature in those directions.

---

## Interpretation

- large positive eigenvalue: strong upward curvature
- small positive eigenvalue: flatter direction
- negative eigenvalue: saddle-like direction
- eigenvalue near zero: nearly flat direction

---

## Why this matters in deep learning

This affects:

- optimization difficulty
- step-size sensitivity
- sharp vs flat directions
- local geometry near minima and saddles

A full Hessian is often too expensive to compute in modern networks, but the conceptual role of Hessian eigenvalues is extremely important.



# 18. AI application: recurrent neural networks and stability

This is a very important application.

In recurrent systems, we repeatedly apply transformations.

A simplified update looks like:

$$
h_t = W h_{t-1}
$$

or in more realistic settings, repeated multiplication by Jacobians appears in backpropagation through time.

If a component aligns with an eigendirection of $W$, repeated multiplication behaves like:

$$
W^k v = \lambda^k v
$$

So:

- if $|\lambda| > 1$, the component can grow rapidly
- if $|\lambda| < 1$, the component can decay rapidly

This gives intuition for exploding and vanishing behavior.

---

## Important technical note

In deep learning, **singular values** are often more directly tied to norm growth and gradient propagation than eigenvalues are. But eigenvalues still provide very useful dynamical intuition.

That distinction matters, but it does not remove the importance of eigen-analysis.



# 19. AI application: representation analysis and interpretability

Eigenvectors and related spectral ideas also appear in neural representation analysis.

Researchers often study:

- covariance matrices of activations
- principal components of hidden states
- low-rank structure of learned representations
- dominant modes of attention patterns

Why?

Because these analyses can reveal:

- major directions of variation
- compressed or low-dimensional structure
- dominant features or patterns
- how information is organized across layers

This is closely connected to representation geometry and mechanistic interpretability.

If you are interested in how concepts are distributed across activation space, spectral tools are among the first mathematical instruments people use to probe that structure.



# 20. AI application: Markov chains and transition matrices

Transition matrices in probabilistic systems also have important eigen-structure.

A common idea is that an eigenvalue of 1 corresponds to a steady-state or stationary behavior.

The other eigenvalues help describe how quickly the system converges toward equilibrium.

This matters in:

- random walks
- ranking algorithms
- graph diffusion
- probabilistic modeling
- some sequential and state-based systems



# 21. Related topic: eigenvalues vs singular values

This is an important related distinction.

## Eigenvalues
Defined for square matrices through:

$$
Av = \lambda v
$$

They describe invariant directions of the transformation.

## Singular values
Defined for any matrix through singular value decomposition (SVD).

They describe how much a matrix stretches space along orthogonal directions.

---

## Why this distinction matters in AI

In machine learning practice:

- eigenvalues are central for covariance matrices, PCA, Hessians, graph methods, dynamical intuition
- singular values are central for conditioning, gradient flow, low-rank approximation, and general rectangular matrices

So eigenvalues and singular values are related, but they are not the same thing.

For symmetric positive semidefinite matrices, the picture becomes especially clean, and eigen-analysis aligns closely with stretch interpretation.



# 22. Common misunderstandings

## Misunderstanding 1: every vector is an eigenvector
No. Most vectors are not.

## Misunderstanding 2: eigenvalue means the length of a vector
No. It is the scaling factor applied to an eigenvector under the transformation.

## Misunderstanding 3: every matrix has real eigenvectors
No. For example, a pure 2D rotation by $90^\circ$ has no real eigenvectors.

## Misunderstanding 4: larger eigenvalue always means "better"
No. The meaning depends on context:
- in PCA: more variance explained
- in dynamics: stronger growth/decay behavior
- in Hessian analysis: stronger curvature
- in graph problems: structural meaning

## Misunderstanding 5: if a vector changes only a little, it is automatically an eigenvector
Not necessarily. For an eigenvector, the output must be an exact scalar multiple of the input.



# 23. Practical computation in Python

In practice, for small to moderate problems, NumPy provides:

- `np.linalg.eig(A)` for general square matrices
- `np.linalg.eigh(A)` for symmetric/Hermitian matrices

For covariance matrices and symmetric matrices, `eigh` is usually preferable because it is designed for that structure.

---

## Example

If `eigvals, eigvecs = np.linalg.eig(A)`, then:

- `eigvals[i]` is the $i$-th eigenvalue
- `eigvecs[:, i]` is the corresponding eigenvector

The columns of the eigenvector matrix correspond to the eigenvalues in the same order.


```python

A = np.array([[4, 2],
              [2, 4]], dtype=float)

eigvals_general, eigvecs_general = np.linalg.eig(A)
eigvals_sym, eigvecs_sym = np.linalg.eigh(A)

print("Using eig:")
print("Eigenvalues:", eigvals_general)
print("Eigenvectors:\n", eigvecs_general)

print("\nUsing eigh (for symmetric matrices):")
print("Eigenvalues:", eigvals_sym)
print("Eigenvectors:\n", eigvecs_sym)
```


# 24. A deeper intuitive summary

A matrix transformation can look complicated in the standard coordinate system.

But if we express vectors in the basis formed by eigenvectors — when such a basis exists — the transformation becomes simple.

Instead of "mixing" coordinates, it just scales them independently.

That is why eigenvectors are so powerful.

They do not merely give us numbers to compute.

They reveal the hidden geometry of a transformation.

---

## In one sentence

Eigenvectors tell us **which directions matter naturally** to the transformation, and eigenvalues tell us **what the transformation does along those directions**.

```python

# Starter code for Exercise 3 and Exercise 4

np.random.seed(42)

# Random symmetric matrix
M = np.random.randn(3, 3)
S = (M + M.T) / 2
eigvals, eigvecs = np.linalg.eigh(S)

print("Random symmetric matrix S:\n", S)
print("\nEigenvalues:", eigvals)
print("\nEigenvectors:\n", eigvecs)

# Simple synthetic 2D data for PCA-style exploration
X = np.random.multivariate_normal(
    mean=[0, 0],
    cov=[[3, 1.5], [1.5, 1]],
    size=200
)

Xc = X - X.mean(axis=0)
C = np.cov(Xc.T)
eigvals_c, eigvecs_c = np.linalg.eigh(C)

print("\nCovariance matrix:\n", C)
print("\nCovariance eigenvalues:", eigvals_c)
print("\nCovariance eigenvectors:\n", eigvecs_c)
```


# 27. Final recap

Let us finish with the full picture.

## Core equation

$$
Av = \lambda v
$$

This means:

- $A$ transforms the vector
- the vector $v$ keeps its line/direction
- the scalar $\lambda$ tells how much it is stretched, shrunk, flipped, or collapsed

---

## What eigenvectors are

They are the special directions that a transformation treats in a pure scaling way.

---

## What eigenvalues are

They are the corresponding scaling factors.

---

## Why we care

Because they reveal the hidden structure of a transformation.

And in AI, understanding hidden structure is everything:

- structure of data
- structure of representations
- structure of optimization landscapes
- structure of dynamics
- structure of graphs and similarities

So eigenvalues and eigenvectors are one of the fundamental bridges between linear algebra and machine learning.

---

## One final sentence to remember

If a matrix is a rule for transforming space, then eigenvectors are the directions that reveal that rule most clearly, and eigenvalues tell us how strongly the rule acts along those directions.
