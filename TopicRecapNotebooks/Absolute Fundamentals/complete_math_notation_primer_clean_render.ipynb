## 1.2 What Is a Function (Formal Definition)

A function is a mapping:

$$
f: X \rightarrow Y
$$

Meaning:

• Domain: $X$  
• Codomain: $Y$  
• For every $x \in X$, there exists exactly one $f(x) \in Y$  

### Important Principle

A function must assign exactly one output per input.

If multiple outputs exist, they must be bundled into a single object (vector, tuple, tensor).

---

## 1.3 Types of Functions in ML

Scalar-valued:

$$
f: \mathbb{R}^n \to \mathbb{R}
$$

Vector-valued:

$$
f: \mathbb{R}^n \to \mathbb{R}^m
$$

Parameterized:

$$
f_\theta(x)
$$

Training finds:

$$
\theta^* = \arg\min_\theta \mathcal{L}(\theta)
$$

---

## 1.4 Function Composition

$$
(g \circ f)(x) = g(f(x))
$$

Deep networks are compositions of many functions.


```

import numpy as np

# Scalar-valued function
def f(x):
    return np.sum(x**2)

x = np.array([1., -2., 3.])
f(x)
```


# 2. Linear Algebra (Geometry of Deep Learning)

Deep learning is linear algebra applied repeatedly.

## 2.1 Vectors and Matrices

Vector:

$$
x \in \mathbb{R}^n
$$

Matrix:

$$
A \in \mathbb{R}^{m \times n}
$$

Matrix entry:

$$
A_{ij}
$$

---

## 2.2 Dot Product

$$
w^T x = \sum_{i=1}^n w_i x_i
$$

Interpretation:

• Weighted sum  
• Similarity measure  

---

## 2.3 Matrix-Vector Multiplication

$$
y = Ax
$$

Component form:

$$
y_i = \sum_{j=1}^n A_{ij} x_j
$$

Interpretation:

Each row of A acts like one neuron.

---

## 2.4 Matrix-Matrix Multiplication

$$
C = AB
$$

$$
C_{ik} = \sum_{j=1}^n A_{ij} B_{jk}
$$

This represents composition of linear transformations.

---

## 2.5 Eigenvalues

$$
Av = \lambda v
$$

Eigenvectors preserve direction.  
Eigenvalues scale magnitude.

---

## 2.6 SVD

$$
A = U \Sigma V^T
$$

Interpretation:

Rotate → Scale → Rotate


```

import numpy as np

A = np.array([[1.,2.,3.],
              [4.,5.,6.]])

x = np.array([10.,20.,30.])

y = A @ x
y
```


# 3. Calculus (Learning via Gradients)

## 3.1 Derivative

$$
f'(x) = \lim_{h \to 0} \frac{f(x+h)-f(x)}{h}
$$

Measures instantaneous rate of change.

---

## 3.2 Gradient

For:

$$
f: \mathbb{R}^n \to \mathbb{R}
$$

$$
\nabla f(x) =
\begin{bmatrix}
\frac{\partial f}{\partial x_1} \\
\vdots \\
\frac{\partial f}{\partial x_n}
\end{bmatrix}
$$

Gradient points in direction of steepest increase.

---

## 3.3 Chain Rule

If:

$$
y = f(g(x))
$$

Then:

$$
\frac{dy}{dx} = f'(g(x)) g'(x)
$$

Backpropagation = repeated chain rule.


```

import torch

x = torch.tensor([1.0, -2.0, 3.0], requires_grad=True)
f = (x**2).sum()
f.backward()

f.item(), x.grad
```


# 4. Probability (Modeling Uncertainty)

## 4.1 Expectation

Discrete:

$$
\mathbb{E}[X] = \sum_x x P(X=x)
$$

Continuous:

$$
\mathbb{E}[X] = \int x p(x) dx
$$

---

## 4.2 Variance

$$
\mathrm{Var}(X) = \mathbb{E}[(X-\mu)^2]
$$

---

## 4.3 Softmax

$$
p_i = \frac{e^{z_i}}{\sum_j e^{z_j}}
$$

Used for categorical distributions.

---

## 4.4 Cross Entropy

$$
- \sum_i y_i \log p_i
$$

---

## 4.5 KL Divergence

$$
D_{KL}(P||Q) = \sum_x P(x) \log \frac{P(x)}{Q(x)}
$$


```

import torch

logits = torch.tensor([2.0,1.0,0.1])
probs = torch.softmax(logits, dim=0)

true_class = torch.tensor(0)
ce = torch.nn.functional.cross_entropy(logits.unsqueeze(0), true_class.unsqueeze(0))

probs, ce.item()
```
