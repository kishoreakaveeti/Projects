## Section 1: Variables and Data Types

## Section 2: Lists and Loops

A **list** is an ordered collection of items. In ML, you'll use lists everywhere:
- List of layer sizes in a network
- List of training losses over time
- List of tokens in a sentence

A **loop** lets you repeat an action for each item in a list.

```python
# Creating lists
layer_sizes = [64, 128, 256, 128, 64]   # a simple neural network architecture
tokens = ["The", "cat", "sat", "on", "the", "mat"]
losses = [2.4, 1.8, 1.2, 0.9, 0.7, 0.65, 0.61]  # training loss going down

print("Layer sizes:", layer_sizes)
print("Number of layers:", len(layer_sizes))
print("First layer:", layer_sizes[0])    # indexing starts at 0!
print("Last layer:", layer_sizes[-1])    # -1 gets the last item
print("Middle layers:", layer_sizes[1:4]) # slicing
```

```python
# for loops — iterate over every item in a list
print("Network architecture:")
for i, size in enumerate(layer_sizes):
    print(f"  Layer {i}: {size} neurons")
```

```python
# while loops — repeat until a condition is False
loss = 2.5
step = 0

print("Simulating training:")
while loss > 1.0:
    loss = loss * 0.75   # loss decreases each step
    step += 1
    print(f"  Step {step}: loss = {loss:.4f}")

print(f"\nTraining complete after {step} steps!")
```

```python
# if / elif / else — conditional logic
current_loss = 0.45

if current_loss < 0.1:
    print("Excellent! Model is well-trained.")
elif current_loss < 0.5:
    print("Good progress, keep training.")
elif current_loss < 1.0:
    print("Model is learning, be patient.")
else:
    print("Loss is high — check your learning rate.")
```

```python
# List comprehensions — a powerful Python shortcut you'll see everywhere
# "normal" way:
doubled = []
for size in layer_sizes:
    doubled.append(size * 2)
print("Doubled (normal):", doubled)

# List comprehension way (same result, one line):
doubled = [size * 2 for size in layer_sizes]
print("Doubled (comprehension):", doubled)

# With a condition:
large_layers = [size for size in layer_sizes if size >= 128]
print("Large layers only:", large_layers)
```

## Section 3: Functions

A **function** is a reusable block of code. Functions are the building blocks of everything in ML — every layer, every activation function, every loss calculation is a function.

In interpretability research, you'll write functions to extract activations, compute similarity, plot circuits, etc.

```python
# Basic function structure
def greet(name):
    """This is a docstring — explains what the function does."""
    message = f"Hello, {name}! Welcome to interpretability research."
    return message

result = greet("Kishore")
print(result)
```

```python
# Functions with multiple parameters and default values
def compute_total_params(num_layers, hidden_size, vocab_size=50257):
    """
    Rough estimate of transformer parameters.
    vocab_size defaults to GPT-2's vocabulary size.
    """
    embedding_params = vocab_size * hidden_size
    attention_params = num_layers * (4 * hidden_size * hidden_size)  # simplified
    mlp_params = num_layers * (8 * hidden_size * hidden_size)        # simplified
    total = embedding_params + attention_params + mlp_params
    return total

# GPT-2 Small: 12 layers, 768 hidden size
gpt2_params = compute_total_params(12, 768)
print(f"GPT-2 Small approximate params: {gpt2_params:,}")
print(f"That's about {gpt2_params / 1_000_000:.0f} million parameters")

# GPT-2 Large: 36 layers, 1280 hidden size
gpt2_large = compute_total_params(36, 1280)
print(f"GPT-2 Large approximate params: {gpt2_large:,}")
```

```python
# THE RELU ACTIVATION FUNCTION
# This is one of the most important functions in deep learning.
# ReLU(x) = max(0, x) — it 'kills' negative values and keeps positive ones.
# Interpretability researchers care a LOT about what gets activated (survives ReLU)

def relu(x):
    """Rectified Linear Unit activation function."""
    return max(0, x)

# Test it on some values
test_values = [-3.0, -0.5, 0.0, 0.5, 2.0, 5.0]
print("Input → ReLU output:")
for val in test_values:
    print(f"  relu({val:5.1f}) = {relu(val):.1f}")
```

```python
# THE SIGMOID FUNCTION
# Maps any number to a value between 0 and 1. 
# Used to represent probabilities.
import math

def sigmoid(x):
    """Sigmoid activation: squashes input to range (0, 1)."""
    return 1 / (1 + math.exp(-x))

print("Input → Sigmoid output (notice all between 0 and 1):")
for val in [-5, -2, -1, 0, 1, 2, 5]:
    print(f"  sigmoid({val:3d}) = {sigmoid(val):.4f}")
```

```python
# THE SOFTMAX FUNCTION
# Converts a list of numbers ("logits") into a probability distribution.
# This is the FINAL step of every language model — it picks the next token!

def softmax(logits):
    """
    Converts raw logits to probabilities.
    All outputs are between 0 and 1, and they sum to 1.
    """
    exp_vals = [math.exp(x) for x in logits]
    total = sum(exp_vals)
    return [e / total for e in exp_vals]

# Imagine a tiny vocab of 5 words, and the model's raw scores
vocab = ["cat", "dog", "sat", "ran", "the"]
logits = [2.0, 1.0, 3.5, 0.5, 1.5]  # model's raw predictions

probs = softmax(logits)

print("Next token probabilities:")
for word, prob in zip(vocab, probs):
    bar = "█" * int(prob * 40)
    print(f"  {word:6s}: {prob:.3f}  {bar}")
print(f"\nSum of all probs: {sum(probs):.6f}  (should be 1.0)")
```

## Section 5: Mini-Project — Build a Single Neuron From Scratch 🧠

Let's put everything together. You're going to implement a **single artificial neuron**.

A neuron:
1. Takes a list of **inputs** (e.g., pixel values, token embeddings)
2. Multiplies each input by a **weight** (how important is this input?)
3. Adds a **bias** (an offset)
4. Passes the result through an **activation function** (e.g., ReLU)

This is the fundamental building block of every neural network, including the transformers that interpretability researchers study.

$$\text{output} = \text{ReLU}\left(\sum_{i} x_i \cdot w_i + b\right)$$

```python
def dot_product(inputs, weights):
    """
    Multiply each input by its corresponding weight and sum them all up.
    This is the core mathematical operation in neural networks.
    In a real model, this is done with matrix multiplication.
    """
    total = 0
    for x, w in zip(inputs, weights):
        total += x * w
    return total


def neuron(inputs, weights, bias, activation="relu"):
    """
    A single artificial neuron.
    
    inputs:     list of input values
    weights:    list of weights (one per input)
    bias:       a single number added to the weighted sum
    activation: which activation function to use
    """
    # Step 1: Weighted sum + bias
    pre_activation = dot_product(inputs, weights) + bias
    
    # Step 2: Apply activation function
    if activation == "relu":
        output = relu(pre_activation)
    elif activation == "sigmoid":
        output = sigmoid(pre_activation)
    else:
        output = pre_activation  # linear (no activation)
    
    return pre_activation, output


# --- Test our neuron ---
# Imagine this neuron detects whether an input contains
# a 'cat-like' feature. Totally made up, but illustrative.

# Input: [fur_score, whisker_score, size_score, bark_score]
cat_input =  [0.9, 0.8, 0.3, 0.0]   # furry, has whiskers, small, no bark
dog_input =  [0.8, 0.1, 0.6, 0.9]   # furry, no whiskers, medium, barks

# Weights: fur and whiskers are strongly positive for 'cat',
#          bark is strongly negative
weights = [0.5, 1.2, -0.3, -2.0]
bias = -0.2

pre_cat, out_cat = neuron(cat_input, weights, bias)
pre_dog, out_dog = neuron(dog_input, weights, bias)

print("=== Cat input ===")
print(f"  Pre-activation (weighted sum + bias): {pre_cat:.3f}")
print(f"  After ReLU: {out_cat:.3f}")
print(f"  Interpretation: {'FIRES (cat-like!)' if out_cat > 0 else 'silent'}")

print("\n=== Dog input ===")
print(f"  Pre-activation (weighted sum + bias): {pre_dog:.3f}")
print(f"  After ReLU: {out_dog:.3f}")
print(f"  Interpretation: {'FIRES' if out_dog > 0 else 'silent (not cat-like)'}")
```

```python
# KEY INSIGHT for interpretability:
# In a real neural network, we don't DESIGN these weights.
# The network LEARNS them automatically from data.
# The question interpretability asks is:
#   "What has each neuron LEARNED to detect?"
#   "Can we UNDERSTAND what concept each neuron represents?"

# Let's experiment: what happens when we change the weights?
print("Experimenting with different weight configurations:")
print("=" * 55)

experiments = [
    {"name": "Original cat detector",    "weights": [0.5, 1.2, -0.3, -2.0], "bias": -0.2},
    {"name": "Bark-sensitive neuron",     "weights": [-0.1, 0.0, 0.0, 3.0],  "bias": -1.0},
    {"name": "Size detector",             "weights": [0.0, 0.0, 2.0, 0.5],   "bias": -0.5},
]

for exp in experiments:
    _, out_cat = neuron(cat_input, exp["weights"], exp["bias"])
    _, out_dog = neuron(dog_input, exp["weights"], exp["bias"])
    print(f"\n{exp['name']}:")
    print(f"  Cat input → {out_cat:.3f}")
    print(f"  Dog input → {out_dog:.3f}")
```
