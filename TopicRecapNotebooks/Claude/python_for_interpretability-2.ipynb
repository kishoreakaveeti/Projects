### 1.3 — Control Flow

Python uses **indentation** instead of curly braces `{}`. This is the biggest syntax change from C#.

````python
# --- IF/ELIF/ELSE ---
loss = 0.45
if loss < 0.1:
    print("Excellent")
elif loss < 0.5:
    print("Training in progress")  # This prints
else:
    print("Loss too high")

# --- FOR LOOP ---
layer_sizes = [768, 512, 256]
for size in layer_sizes:
    print(f"Layer size: {size}")

# --- enumerate: index + value together ---
# C# style: for(int i=0; i<list.Count; i++)
# Python:
for i, size in enumerate(layer_sizes):
    print(f"Layer {i}: {size}")

# --- range ---
for epoch in range(5):
    print(f"Epoch {epoch}")

# --- WHILE ---
loss = 1.0
steps = 0
while loss > 0.5:
    loss *= 0.7
    steps += 1
print(f"Converged in {steps} steps, final loss: {loss:.4f}")
````

No return type declarations. Default arguments are easy. `**kwargs` (keyword arguments) are common in ML libraries.

````python
# Basic function
def relu(x):
    """ReLU activation: max(0, x). Triple quotes = Python docstring."""
    return max(0, x)

print(relu(-3))  # 0
print(relu(5))   # 5

# Default arguments
def create_layer(input_dim, output_dim, activation="relu", dropout=0.1):
    return {"in": input_dim, "out": output_dim,
            "activation": activation, "dropout": dropout}

layer1 = create_layer(768, 256)              # uses defaults
layer2 = create_layer(256, 64, dropout=0.2)  # override one
print(layer1)
print(layer2)

# Multiple return values (returns a tuple)
def train_step(loss_value):
    accuracy = 1.0 - loss_value * 0.3
    improved = loss_value < 0.5
    return loss_value, accuracy, improved

loss, acc, did_improve = train_step(0.4)
print(f"Loss: {loss}, Accuracy: {acc:.3f}, Improved: {did_improve}")

# Lambda — anonymous one-liner
sigmoid = lambda x: 1 / (1 + 2.718 ** (-x))
print(f"Sigmoid(0) = {sigmoid(0):.4f}")
print(f"Sigmoid(2) = {sigmoid(2):.4f}")
````

### 1.6 — Classes and Object-Oriented Python

Familiar from C#. Key difference: `self` is Python's `this`. In PyTorch, everything is a class.

````python
class NeuralNetwork:
    framework = "PyTorch"  # class variable

    def __init__(self, name, layer_sizes):
        self.name = name
        self.layer_sizes = layer_sizes
        self.n_layers = len(layer_sizes)

    def total_parameters(self):
        total = 0
        for i in range(len(self.layer_sizes) - 1):
            total += self.layer_sizes[i] * self.layer_sizes[i+1]
            total += self.layer_sizes[i+1]
        return total

    def summary(self):
        print(f"Model: {self.name} | Layers: {self.layer_sizes}")
        print(f"Total parameters: {self.total_parameters():,}")

    def __repr__(self):
        return f"NeuralNetwork('{self.name}', layers={self.n_layers})"


gpt_small = NeuralNetwork("GPT-Small", [768, 768, 768, 50257])
gpt_small.summary()
print(gpt_small)

# Inheritance
class TransformerModel(NeuralNetwork):
    def __init__(self, name, d_model, n_heads, n_layers):
        layer_sizes = [d_model] * n_layers
        super().__init__(name, layer_sizes)
        self.d_model = d_model
        self.n_heads = n_heads

    def head_dim(self):
        return self.d_model // self.n_heads

    def summary(self):
        super().summary()
        print(f"Heads: {self.n_heads}, Head dim: {self.head_dim()}")


gpt2 = TransformerModel("GPT-2", d_model=768, n_heads=12, n_layers=12)
gpt2.summary()
````

### 4.1 — Building Neural Networks with torch.nn

Build models by subclassing `nn.Module` — exactly like the OOP pattern from Part 1.

````python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleTransformerBlock(nn.Module):
    """
    A simplified Transformer block.
    Contains: LayerNorm -> Attention -> Residual + LayerNorm -> MLP -> Residual
    """
    def __init__(self, d_model=64, n_heads=4, d_ff=256, dropout=0.1):
        super().__init__()
        self.n_heads = n_heads
        self.ln1 = nn.LayerNorm(d_model)
        self.ln2 = nn.LayerNorm(d_model)
        self.qkv_proj = nn.Linear(d_model, 3 * d_model, bias=False)
        self.out_proj = nn.Linear(d_model, d_model, bias=False)
        self.mlp = nn.Sequential(
            nn.Linear(d_model, d_ff),
            nn.GELU(),
            nn.Linear(d_ff, d_model),
            nn.Dropout(dropout)
        )
        self.dropout = nn.Dropout(dropout)

    def attention(self, x):
        B, T, C = x.shape
        head_dim = C // self.n_heads
        qkv = self.qkv_proj(x)
        q, k, v = qkv.chunk(3, dim=-1)
        q = q.view(B, T, self.n_heads, head_dim).transpose(1, 2)
        k = k.view(B, T, self.n_heads, head_dim).transpose(1, 2)
        v = v.view(B, T, self.n_heads, head_dim).transpose(1, 2)
        scale = head_dim ** -0.5
        scores = (q @ k.transpose(-2, -1)) * scale
        mask = torch.triu(torch.ones(T, T, device=x.device), diagonal=1).bool()
        scores = scores.masked_fill(mask, float('-inf'))
        weights = F.softmax(scores, dim=-1)
        weights = self.dropout(weights)
        out = weights @ v
        out = out.transpose(1, 2).contiguous().view(B, T, C)
        return self.out_proj(out), weights

    def forward(self, x):
        attn_out, attn_weights = self.attention(self.ln1(x))
        x = x + attn_out          # residual stream!
        x = x + self.mlp(self.ln2(x))  # residual stream again!
        return x, attn_weights


block = SimpleTransformerBlock(d_model=64, n_heads=4)
x = torch.randn(2, 10, 64)
output, weights = block(x)
print("Input shape: ", x.shape)
print("Output shape:", output.shape)   # same as input — residual stream!
print("Attn weights:", weights.shape)  # (batch, heads, seq, seq)
total = sum(p.numel() for p in block.parameters())
print(f"Parameters: {total:,}")
````

### 4.2 — A Complete Training Loop

````python
import torch, torch.nn as nn, matplotlib.pyplot as plt

class SimpleClassifier(nn.Module):
    def __init__(self, input_dim=10, hidden_dim=32, n_classes=2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, n_classes)
        )
    def forward(self, x):
        return self.net(x)

torch.manual_seed(42)
X = torch.randn(1000, 10)
y = (X.sum(dim=1) > 0).long()
X_train, X_val = X[:800], X[800:]
y_train, y_val = y[:800], y[800:]

model     = SimpleClassifier()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

train_losses, val_losses = [], []
for epoch in range(50):
    model.train()
    optimizer.zero_grad()
    loss = criterion(model(X_train), y_train)
    loss.backward()
    optimizer.step()
    train_losses.append(loss.item())

    model.eval()
    with torch.no_grad():
        val_loss = criterion(model(X_val), y_val)
        val_losses.append(val_loss.item())

model.eval()
with torch.no_grad():
    acc = (model(X_val).argmax(1) == y_val).float().mean()
    print(f"Validation Accuracy: {acc:.1%}")

plt.figure(figsize=(8, 3))
plt.plot(train_losses, label='Train'); plt.plot(val_losses, label='Val', linestyle='--')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.title('Training Loss')
plt.legend(); plt.grid(alpha=0.3); plt.show()
````

### 5.2 — einops: Clean Tensor Operations

You'll see einops in virtually every interpretability codebase. It makes tensor reshaping readable.

````python
from einops import rearrange, reduce, repeat
import torch

x = torch.randn(2, 10, 64)  # (batch, seq, d_model)

# Confusing old way:
B, T, C = x.shape
old = x.reshape(B, T, 4, C//4).transpose(1, 2)
print("Old way:", old.shape)

# Readable einops way:
new = rearrange(x, 'b t (h d) -> b h t d', h=4)
print("einops:  ", new.shape)
print("Same result:", torch.allclose(old, new))

# More examples
merged   = rearrange(new, 'b h t d -> b t (h d)')
print("Merged:", merged.shape)

avg      = reduce(new, 'b h t d -> b t d', 'mean')
print("Avg over heads:", avg.shape)

pos_emb  = torch.randn(10, 64)
batched  = repeat(pos_emb, 't d -> b t d', b=8)
print("Repeated to batch:", batched.shape)
````

## Part 6: Mini Research Exercise — Investigating Attention Heads

A real interpretability task: do different attention heads in GPT-2 specialize?

We'll measure:
1. How much each head attends to the **previous token** (local)
2. How much each head attends to the **first token** (global)

````python
import torch, numpy as np, matplotlib.pyplot as plt
from transformers import AutoTokenizer, AutoModelForCausalLM

def get_attention_patterns(text):
    inputs = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        out = model(**inputs, output_attentions=True)
    attn = torch.stack(out.attentions)[:, 0]  # (n_layers, n_heads, seq, seq)
    return attn.numpy()

text = "The quick brown fox jumps over the lazy dog"
attn = get_attention_patterns(text)
n_layers, n_heads, seq_len, _ = attn.shape
print(f"Attention shape: {attn.shape}")
````

````python
prev_token_attn  = np.zeros((n_layers, n_heads))
first_token_attn = np.zeros((n_layers, n_heads))

for layer in range(n_layers):
    for head in range(n_heads):
        a = attn[layer, head]
        prev_token_attn[layer, head]  = np.mean([a[i, i-1] for i in range(1, seq_len)])
        first_token_attn[layer, head] = np.mean(a[:, 0])

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
im1 = axes[0].imshow(prev_token_attn,  cmap='hot', aspect='auto')
im2 = axes[1].imshow(first_token_attn, cmap='hot', aspect='auto')
for ax, im, title in zip(axes, [im1,im2],
    ['Avg Attention to PREVIOUS Token', 'Avg Attention to FIRST Token']):
    plt.colorbar(im, ax=ax)
    ax.set_xlabel('Head'); ax.set_ylabel('Layer'); ax.set_title(title)
plt.suptitle('GPT-2 Head Specialization', fontsize=13)
plt.tight_layout(); plt.show()

best_prev  = np.unravel_index(prev_token_attn.argmax(),  prev_token_attn.shape)
best_first = np.unravel_index(first_token_attn.argmax(), first_token_attn.shape)
print(f"Most local head:  Layer {best_prev[0]},  Head {best_prev[1]}")
print(f"Most global head: Layer {best_first[0]}, Head {best_first[1]}")
print("Different heads DO specialize — this is real interpretability!")
````
