
# Deep Neural Networks: Formal Representation Learning Perspective

This notebook formalizes the conceptual description of deep learning systems:

- Neural networks as parameterized function families
- Layer-wise representation transformation
- Role of nonlinearity
- Inductive bias vs unrestricted parameterization
- Optimization and backpropagation
- Geometric interpretation of representation learning
- Extension to Transformers

All equations are written using MathJax-friendly syntax.



## 1. Neural Networks as Parameterized Function Families

A neural network defines a function:

$$
f(x; \theta)
$$

Where:

- $x$ is the input
- $\theta$ represents all learnable parameters (weights and biases)

The architecture defines the **functional form**.
Training determines the specific parameter instantiation $\theta$.



## 2. Deep Networks as Compositions of Functions

A deep network is a composition:

$$
f(x) = f_L \circ f_{L-1} \circ \dots \circ f_1 (x)
$$

Each layer takes the form:

$$
f_\ell(h) = \sigma(W_\ell h + b_\ell)
$$

Thus:

$$
h_1 = \sigma(W_1 x + b_1)
$$

$$
h_2 = \sigma(W_2 h_1 + b_2)
$$

$$
\vdots
$$

$$
y = W_L h_{L-1} + b_L
$$



## 3. Why Nonlinearity is Necessary

If all layers were linear:

$$
h_1 = W_1 x
$$

$$
h_2 = W_2 h_1 = W_2 W_1 x
$$

Stacking linear maps gives:

$$
W_L W_{L-1} \dots W_1 x = W x
$$

Which is still linear.

Therefore, without nonlinear activation $\sigma$, deep networks collapse to a single linear transformation.



## 4. Representation Learning Formalized

A representation is a mapping:

$$
\phi(x)
$$

Such that the final task becomes simpler:

$$
f(x) = g(\phi(x))
$$

Deep networks attempt to learn $\phi(x)$ such that:

$$
g(\phi(x))
$$

is simple (often linear).



## 5. Inductive Bias vs Unlimited Parameterization

The goal of deep learning is NOT to make everything variable.

Instead, architectures introduce structured parameterization.

Fully Connected Layer parameter count:

$$
(HWC_{in}) \cdot M + M
$$

Convolutional Layer parameter count:

$$
(C_{out} \cdot C_{in} \cdot k_h \cdot k_w) + C_{out}
$$

Convolution introduces:

- Local connectivity
- Weight sharing
- Translation equivariance



## 6. Optimization Objective

Training minimizes expected loss:

$$
\min_{\theta} \mathbb{E}_{(x,y)} [ \mathcal{L}(f(x;\theta), y) ]
$$

Gradient descent update rule:

$$
\theta \leftarrow \theta - \eta \nabla_\theta \mathcal{L}
$$

Backpropagation applies the chain rule:

$$
\frac{\partial \mathcal{L}}{\partial W_\ell}
=
\frac{\partial \mathcal{L}}{\partial h_\ell}
\frac{\partial h_\ell}{\partial W_\ell}
$$



## 7. Geometric View of Representation Learning

Let the input distribution lie on a manifold $\mathcal{M}$.

Each layer applies:

$$
h_{\ell+1} = \sigma(W_\ell h_\ell)
$$

This can be viewed as warping the manifold to increase linear separability.

This idea is sometimes referred to as the manifold flattening hypothesis.



## 8. Extension to Transformers

Transformer layer:

$$
h_{\ell+1} =
\text{FFN}(\text{Attention}(h_\ell))
$$

Self-attention:

$$
\text{Attention}(Q,K,V) =
\text{softmax}\left(\frac{QK^T}{\sqrt{d}}\right)V
$$

Where:

$$
Q = XW_Q, \quad K = XW_K, \quad V = XW_V
$$



## 9. Unified Statement

Deep neural networks are structured parameterized function families composed of alternating linear transformations and nonlinearities.

Their purpose is to learn hierarchical representations that transform complex input distributions into geometrically simpler forms for the target task.
