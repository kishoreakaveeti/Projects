## 🧠 1. Motivation

Deep neural networks (especially Transformers) suffer from:

- Activation drift across layers
- Instability in residual connections
- Exploding or vanishing gradients
- Sensitivity to initialization

Layer Normalization stabilizes training by normalizing activations at each layer.



## 📐 2. Mathematical Definition

Let a hidden representation for one token be:

x = [x₁, x₂, ..., x_d] ∈ ℝᵈ

### Step 1 — Mean

μ = (1/d) Σ xᵢ

### Step 2 — Variance

σ² = (1/d) Σ (xᵢ − μ)²

### Step 3 — Normalize

x̂ᵢ = (xᵢ − μ) / √(σ² + ε)

### Step 4 — Scale and Shift

yᵢ = γᵢ x̂ᵢ + βᵢ

Where γ and β are learnable parameters.



## 📊 3. Geometric Interpretation

After normalization:

- Mean = 0
- Variance = 1
- Stable magnitude
- Reduced scale sensitivity

Vectors lie on a hypersphere in feature space.



## 🧮 4. Manual Numerical Example

Let:

x = [2, 4, 6]

Mean = 4  
Variance = 8/3  
Std ≈ 1.633  

Normalized = [-1.2247, 0, 1.2247]


```python

import numpy as np

def layer_norm(x, gamma, beta, eps=1e-5):
    mean = np.mean(x)
    var = np.var(x)
    x_hat = (x - mean) / np.sqrt(var + eps)
    y = gamma * x_hat + beta
    return y, mean, var

# Test example
x = np.array([2.0, 4.0, 6.0])
gamma = np.ones(3)
beta = np.zeros(3)

y, mean, var = layer_norm(x, gamma, beta)

print("Input:", x)
print("Mean:", mean)
print("Variance:", var)
print("Output:", y)
```


## 🔬 5. Multi-Token Simulation (Transformer Style)

LayerNorm operates across the last dimension (hidden dimension).


```python

# Simulating (batch=2, sequence=3, hidden_dim=4)

x = np.random.randn(2, 3, 4)

def layer_norm_batch(x, eps=1e-5):
    mean = np.mean(x, axis=-1, keepdims=True)
    var = np.var(x, axis=-1, keepdims=True)
    x_hat = (x - mean) / np.sqrt(var + eps)
    return x_hat

normalized = layer_norm_batch(x)

print("Original shape:", x.shape)
print("Normalized shape:", normalized.shape)
```


## 🏗 6. Position in Transformer Block

Two variants:

### Post-LN (Original)

x + Sublayer(x) → LayerNorm

### Pre-LN (Modern LLMs)

LayerNorm(x) → Sublayer → +x

Pre-LN improves gradient flow in very deep networks.


```python

import torch
import torch.nn as nn

layer_norm = nn.LayerNorm(4)

x = torch.randn(2, 3, 4)
output = layer_norm(x)

print("Output shape:", output.shape)
```


## 🎯 7. Key Takeaways

LayerNorm:

- Normalizes across feature dimension
- Independent of batch size
- Stabilizes deep residual networks
- Essential for Transformers
- Improves gradient flow
