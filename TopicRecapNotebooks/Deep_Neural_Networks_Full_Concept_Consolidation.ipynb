# 1. Universal Approximation Theorem

Consider a one-hidden-layer neural network:

$$
\hat{f}(x) = \sum_{i=1}^{m} a_i \sigma(w_i^T x + b_i) + c
$$

Where:

- $x \in \mathbb{R}^d$
- $\sigma$ is a non-polynomial activation function
- $m$ is the number of hidden neurons

### Theorem (Classical Form)

Let $K \subset \mathbb{R}^d$ be compact and let $f:K \to \mathbb{R}$ be continuous.

For any $\varepsilon > 0$, there exists parameters such that:

$$
\sup_{x \in K} |f(x) - \hat{f}(x)| < \varepsilon
$$

This guarantees **existence of approximation**, not trainability.



# 2. Multi-Dimensional Inputs and Embeddings

When we write:

$$
f(x)
$$

we mean:

$$
f(x_1, x_2, \dots, x_d)
$$

A neuron computes:

$$
\sigma(w^T x + b) = \sigma\left(\sum_{j=1}^d w_j x_j + b\right)
$$

In embedding space (e.g., $x \in \mathbb{R}^{768}$), the network learns partitions of high-dimensional space.

Each neuron defines a hyperplane:

$$
w^T x + b = 0
$$

ReLU networks partition input space into polyhedral regions.



# 3. Continuous vs Discontinuous Functions

If activation $\sigma$ is continuous, then the network output is continuous.

Thus:

- Exact discontinuities cannot be represented.
- Uniform approximation of a step function is impossible.
- $L^p$ approximation is possible.

Example step function:

$$
f(x) =
\begin{cases}
0 & x < 0 \\
1 & x \ge 0
\end{cases}
$$

Sigmoid approximation:

$$
f_k(x) = \frac{1}{1 + e^{-kx}}
$$

As $k \to \infty$, $f_k(x)$ approaches the step function.



# 4. Conditional Jumps and Outliers

General generative model:

$$
y = g(x) + h(x) + \varepsilon
$$

Where:

- $g(x)$ = smooth baseline
- $h(x)$ = conditional structure (piecewise)
- $\varepsilon$ = noise

Neural networks approximate:

$$
\mathbb{E}[y \mid x]
$$

Rare outliers influence training based on the loss function.



# 5. Decision Boundaries

Binary classifier:

$$
\hat{y} =
\begin{cases}
1 & f(x) > 0 \\
0 & \text{otherwise}
\end{cases}
$$

Decision boundary:

$$
f(x) = 0
$$

ReLU networks are piecewise linear:

Inside each region:

$$
f(x) = A_r x + c_r
$$

Deep networks naturally form complex decision boundaries during training.

Parameter update:

$$
\theta_{t+1} = \theta_t - \eta \nabla L(\theta_t)

Gradual weight updates can cause abrupt representational reorganization.

# 8. Activation Functions

## Sigmoid

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Derivative:

$$
\sigma'(x) = \sigma(x)(1-\sigma(x))
$$

Suffers from vanishing gradients.

---

## Tanh

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

Zero-centered, but still saturates.

---

## ReLU

$$
\text{ReLU}(x) = \max(0,x)
$$

Derivative:

$$
\text{ReLU}'(x) =
\begin{cases}
1 & x>0 \\
0 & x<0
\end{cases}
$$

Enables stable gradient flow and piecewise linear partitioning.



# 9. Gradient Flow Across Layers

Backpropagation multiplies derivatives:

$$
\frac{\partial L}{\partial x} =
\frac{\partial L}{\partial f_L}
\prod_{l=1}^L \sigma'(z_l) W_l
$$

If $|\sigma'| < 1$ consistently → vanishing gradient.

ReLU maintains gradient magnitude in active region.

This explains why ReLU enabled deep learning breakthroughs.
