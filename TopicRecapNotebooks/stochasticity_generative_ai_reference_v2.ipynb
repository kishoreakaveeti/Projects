
# Stochasticity in Deep Learning & Generative AI  
*(Clean MathJax-Compatible Version)*

This notebook provides a mathematically precise explanation of **stochasticity** and its role in deep neural networks and generative AI.



## 1. Deterministic vs Stochastic Systems

### Deterministic Mapping

$$
y = f(x)
$$

Same input → same output.

### Stochastic Mapping

$$
y \sim p(y \mid x)
$$

This means: *y is sampled from the probability distribution p(y | x).*

Instead of producing a single value, the model defines a distribution over possible outputs.



## 2. Random Variables

A random variable is a function:

$$
X : \Omega \rightarrow \mathbb{R}
$$

Where:

- $\Omega$ = sample space  
- $X$ assigns numerical values to outcomes  

For continuous distributions:

$$
\int p(x) \, dx = 1
$$



## 3. Expectation and Variance

### Expectation

$$
\mathbb{E}[X] = \int x \, p(x) \, dx
$$

### Variance

$$
\mathrm{Var}(X) = \mathbb{E}[(X - \mathbb{E}[X])^2]
$$

Variance measures uncertainty.



## 4. Stochastic Gradient Descent

Full gradient:

$$
\nabla_\theta L(\theta)
$$

Mini-batch estimate:

$$
\nabla_\theta L_B(\theta)
$$

The mini-batch introduces randomness (noise) into optimization.



## 5. Traditional Autoencoder (Deterministic)

Encoder:

$$
z = f_\theta(x)
$$

Decoder:

$$
\hat{x} = g_\phi(z)
$$

The mapping is deterministic.



## 6. Variational Autoencoder (Introducing Stochasticity)

Instead of a fixed latent vector:

$$
z \sim q_\phi(z \mid x)
$$

Common assumption:

$$
q_\phi(z \mid x) = \mathcal{N}(\mu(x), \sigma^2(x))
$$

Reparameterization trick:

$$
z = \mu + \sigma \epsilon
$$

where

$$
\epsilon \sim \mathcal{N}(0,1)
$$



## 7. ELBO Objective

The Evidence Lower Bound:

$$
\mathcal{L}(\theta,\phi; x) =
\mathbb{E}_{q_\phi(z|x)}[\log p_\theta(x|z)]
- D_{KL}(q_\phi(z|x) \| p(z))
$$

Where:

- First term: reconstruction likelihood  
- Second term: KL divergence regularizer  

Usually:

$$
p(z) = \mathcal{N}(0,1)
$$



## 8. Diffusion Model Forward Process

Noise is added progressively:

$$
x_t = \sqrt{\alpha_t} x_{t-1} + \sqrt{1-\alpha_t} \epsilon
$$

where

$$
\epsilon \sim \mathcal{N}(0,1)
$$

The model learns to reverse this stochastic corruption process.



## 9. Core Insight

Deterministic models learn:

$$
f(x)
$$

Stochastic generative models learn:

$$
p(x)
$$

That shift — from function learning to distribution learning — is the foundation of modern generative AI.

# Mini Demonstration: Reparameterization Trick


```

import numpy as np

rng = np.random.default_rng(0)

mu = np.array([0.5, -1.0, 0.0])
sigma = np.array([0.2, 0.5, 1.0])

eps = rng.normal(0, 1, size=(10, 3))
z_samples = mu + sigma * eps

print("mu:", mu)
print("sigma:", sigma)
print("\nFirst 5 latent samples:")
print(z_samples[:5])

print("\nSample mean:", z_samples.mean(axis=0))
print("Sample std:", z_samples.std(axis=0))
```
