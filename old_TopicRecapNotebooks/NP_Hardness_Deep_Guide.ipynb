## 1. Efficient vs Inefficient Computation

Let input size be $n$. We analyze how runtime grows as a function of $n$.

### Polynomial Time (Efficient)

$$
T(n) = O(n^k)
$$

for some constant $k$. Examples:

- $O(n)$
- $O(n^2)$
- $O(n^3)$
- $O(n^{10})$ (still polynomial)

### Exponential Time (Infeasible Scaling)

$$
T(n) = O(2^n), \quad O(3^n), \quad O(n!)
$$

### Numerical Comparison

If $n=100$:

- $n^3 = 10^6$
- $2^{100} \approx 1.27 \times 10^{30}$

Even at $10^{12}$ operations per second:

$$
\frac{2^{100}}{10^{12}} \approx 1.27 \times 10^{18} \text{ seconds}
$$

which is approximately $4 \times 10^{10}$ years.

This illustrates why exponential scaling is computationally infeasible.


## 2. Decision vs Optimization Problems

Complexity theory classifies **decision problems** (yes/no questions).

### Example: Traveling Salesman Problem (TSP)

Optimization version:
> Find the shortest tour.

Decision version:

$$
\exists \text{ tour } \pi \text{ such that } \text{length}(\pi) \le B ?
$$

Hardness results are typically defined for decision versions.


## 3. Class P

A decision problem is in **P** if it can be solved in polynomial time:

$$
T(n) = O(n^k)
$$

Examples:

- Shortest path
- Minimum spanning tree
- Matrix multiplication
- Linear regression (convex case)


## 4. Class NP

A problem is in **NP** if YES instances have certificates verifiable in polynomial time.

Formally:

$$
x \in L \iff \exists c \text{ with } |c| \le \text{poly}(|x|)
$$

such that a verifier $V(x,c)$ runs in polynomial time and returns 1.

### Example: Subset Sum

Given numbers $a_1, \dots, a_n$ and target $T$, does

$$
\exists S \subseteq \{1,\dots,n\} : \sum_{i \in S} a_i = T ?
$$

Certificate: subset $S$.

Verification cost:

$$
O(n)
$$


## 5. Reductions

We say problem $A$ reduces to problem $B$ if:

$$
A \le_p B
$$

meaning there exists a polynomial-time function $f$ such that:

$$
x \in A \iff f(x) \in B.
$$

If $B$ can solve $f(x)$ efficiently, then $A$ can be solved efficiently.

Reductions are how hardness is transferred between problems.


## 6. NP-hard

A problem $B$ is NP-hard if:

$$
\forall A \in NP, \quad A \le_p B.
$$

This means $B$ is at least as hard as every problem in NP.

Important:

- NP-hard does NOT mean impossible.
- It means no known polynomial-time algorithm for worst-case instances.


## 7. NP-complete

A problem is NP-complete if:

1. It is in NP.
2. It is NP-hard.

These are the hardest problems within NP.

If any NP-complete problem is solved in polynomial time, then:

$$
P = NP
$$

which remains an open problem.


## 8. Classic NP-complete Problems

### 8.1 SAT

Given Boolean formula $\varphi(x_1, \dots, x_n)$:

$$
\exists (x_1,\dots,x_n) \text{ such that } \varphi = TRUE?
$$

### 8.2 3-SAT

$$
\varphi = (l_{11} \vee l_{12} \vee l_{13}) \wedge \dots \wedge (l_{m1} \vee l_{m2} \vee l_{m3})
$$

Each $l_{ij}$ is $x_k$ or $\neg x_k$.

### 8.3 Clique

Given graph $G=(V,E)$:

$$
\exists C \subseteq V, \quad |C| \ge k
$$

such that

$$
\forall u \ne v \in C, \quad (u,v) \in E
$$

### 8.4 Vertex Cover

$$
\exists S \subseteq V, \quad |S| \le k
$$

such that

$$
\forall (u,v) \in E, \quad u \in S \text{ or } v \in S
$$

### 8.5 Knapsack

$$
\exists S \subseteq \{1..n\} :
\sum_{i \in S} w_i \le W \quad \text{and} \quad \sum_{i \in S} v_i \ge V ?
$$

### 8.6 TSP (decision)

$$
\exists \text{ tour } \pi \text{ such that } \text{length}(\pi) \le B?
$$


## 9. Combinatorial Explosion

In subset selection problems, each of $m$ elements can be:

- Selected
- Not selected

Total combinations:

$$
2^m
$$

Example:

If $m = 10^7$:

$$
2^{10^7}
$$

is astronomically large.

This combinatorial structure underlies many NP-hard problems.


## 10. Sparse Model Selection is NP-hard

Consider sparse regression:

$$
\min_{\beta} \|y - X\beta\|_2^2
\quad \text{s.t.} \quad \|\beta\|_0 \le k
$$

Here:

$$
\|\beta\|_0 = \# \{ j : \beta_j \ne 0 \}
$$

This requires searching over subsets of features:

$$
2^d \text{ possible subsets}
$$

Therefore it is NP-hard.

Relaxation:

$$
\min_{\beta} \|y - X\beta\|_2^2 + \lambda \|\beta\|_1
$$

This is convex and solvable in polynomial time.


## 11. Recognizing NP-hardness in Research

Warning signs:

- Binary variables: $z_i \in \{0,1\}$
- Subset selection constraints: $\|\theta\|_0 \le k$
- Integer programming formulations
- Combinatorial search over exponentially many structures

If your problem reduces to selecting a minimal subset under constraints, it is likely NP-hard.


## 12. Final Mental Model

- **P**: Solvable in polynomial time.
- **NP**: Solutions verifiable in polynomial time.
- **NP-complete**: Hardest problems in NP.
- **NP-hard**: At least as hard as NP-complete problems.

NP-hard does not mean impossible.

It means:

> No known polynomial-time algorithm exists for solving all instances exactly, and worst-case complexity appears exponential.
